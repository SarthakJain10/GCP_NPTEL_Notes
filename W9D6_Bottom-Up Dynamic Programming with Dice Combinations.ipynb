# Bottom-Up Dynamic Programming with Dice Combinations

This lecture introduces **Bottom-Up Dynamic Programming (DP)**, contrasting it with the top-down (memoized recursion) approach. The concept is illustrated using the "Dice Combinations" problem from the CSES problem set.

## Introduction to Bottom-Up Dynamic Programming
* **Bottom-Up DP** is a method of solving problems by using tables (arrays) instead of recursive functions.
* Unlike top-down DP, which starts with the target problem and breaks it down, the bottom-up approach starts with the smallest sub-problems (**base cases**) and builds up to the final solution.
* It is often described as "clever brute force" or "filling a table" in a specific order.

## Problem Statement: Dice Combinations
The task is to count the number of distinct sequences of dice throws that sum up to a given integer `n`.
* **Constraints:** Each dice throw results in an outcome between 1 and 6.
* **Nature of Problem:** This is a **pure counting question**, not a probability question. We are counting the total number of valid sequences.

### Examples
* **n = 1:** Only one way: `(1)`.
* **n = 2:** Two ways: `(1, 1)` and `(2)`.
* **n = 3:** Four ways: `(3)`, `(2, 1)`, `(1, 2)`, and `(1, 1, 1)`.
* **n = 4:** Eight ways. This is calculated by looking at the last dice roll:
    * If the last roll is `1`, the remaining sum is `3` (4 ways to get 3).
    * If the last roll is `2`, the remaining sum is `2` (2 ways to get 2).
    * If the last roll is `3`, the remaining sum is `1` (1 way to get 1).
    * If the last roll is `4`, the remaining sum is `0` (1 way to get 0).
    * Total: $4 + 2 + 1 + 1 = 8$.

## Developing the Recurrence Relation
To solve for a sum `n`, we categorize the solution space based on the **last dice roll** (represented as `k`).
* The last roll `k` can be any value from $\{1, 2, 3, 4, 5, 6\}$.
* If the last roll is `k`, the previous rolls must have summed to `n - k`.
* **Recurrence:** `Ways(n) = Ways(n-1) + Ways(n-2) + Ways(n-3) + Ways(n-4) + Ways(n-5) + Ways(n-6)`.
* **Base Case:** `Ways(0) = 1`. This represents the "empty sequence" used when a single roll exactly matches the target sum.
* **Edge Cases:** For any `n < 0`, `Ways(n) = 0`.


## Bottom-Up Implementation Strategy
Instead of calling a function recursively, we use a one-dimensional array (table) to store results.
1.  Initialize an array `DP` of size `n + 1`.
2.  Set the base case: `DP[0] = 1`.
3.  Iterate from `i = 1` to `n`.
4.  For each `i`, calculate the sum of the previous six entries: `DP[i-1]` through `DP[i-6]`.
5.  Ensure that indices less than 0 are treated as 0 (sanity checks for small `i`).

### Pseudocode Logic
```python
function countDiceCombinations(n):
    create array DP of size n + 1
    initialize all entries in DP to 0
    
    # Base Case
    DP[0] = 1
    
    # Fill the table from left to right
    for i from 1 up to n:
        for dice_value from 1 up to 6:
            if i - dice_value >= 0:
                DP[i] = (DP[i] + DP[i - dice_value]) modulo 10^9 + 7
                
    return DP[n]
```

## Comparison: Top-Down vs. Bottom-Up
| Feature | Top-Down (Memoization) | Bottom-Up (Tabulation) |
| :--- | :--- | :--- |
| **Logic** | Recursive, starts from `n`. | Iterative, starts from base case. |
| **Computation** | Only computes sub-problems needed for the answer. | Typically computes the full table. |
| **Memory** | Uses recursion stack space. | Can be easier to optimize for memory. |
| **Order** | Order of sub-problems is handled by recursion. | Order must be explicitly defined by the programmer. |

### Memory Optimization Note
In the Dice Combinations problem, we only ever need the **previous six entries** to calculate the current one. In a bottom-up approach, we could theoretically use only a small constant-sized array (size 6 or 7) and overwrite it, reducing space complexity. This is harder to implement in a top-down recursive style.

## Key Takeaways
* **DP Philosophy:** Focus on two questions: **What** do you want to store, and **how** do you compute it using previous values?
* **Semantics:** If a recurrence isn't providing enough information, you may need to "strengthen the semantics" by storing more data in each table entry.
* **Order of Evaluation:** In bottom-up DP, you must ensure that all dependencies (sub-problems) are solved and stored in the table before the current problem is processed.
* **Large Numbers:** Problems like this often require the use of a modulo (e.g., $10^9 + 7$) because the number of combinations grows exponentially.
