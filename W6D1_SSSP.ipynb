# Shortest Path Algorithms: Overview and Breadth-First Search (BFS)

This lecture provides a high-level overview of the **Shortest Path Problem** in weighted and directed graphs. It explores different scenarios based on edge weights and cycles, focusing specifically on how **Breadth-First Search (BFS)** serves as the foundation for solving these problems in unweighted graphs.

---

## 1. Introduction to Shortest Paths
The primary goal of shortest path algorithms is to determine the fastest or "cheapest" way to navigate from one vertex to another in a graph.

### Key Graph Characteristics
* **Weighted Graphs:** Every edge is mapped to a numerical value (weight). The cost of a path is the sum of these weights.
* **Directed Graphs:** Edges have a specific orientation; an edge from vertex $X$ to $Y$ is distinct from an edge from $Y$ to $X$.
* **Distance:** The length of the shortest path connecting a source vertex $S$ to a target vertex $T$.

### Problem Variants
1.  **Single Source Shortest Path (SSSP):** Finding the distance from one fixed source vertex to every other vertex in the graph.
2.  **All-Pairs Shortest Path (APSP):** Finding the distance between every possible pair of vertices in the graph.
3.  **Negative Weights/Cycles:** A specialized scenario where edges can have negative values. A **negative weight cycle** is a path starting and ending at the same vertex where the total sum of edge weights is less than zero.

---

## 2. Path Terminology
The lecture distinguishes between different types of traversals to ensure clarity in algorithmic logic:

* **Path:** A sequence of vertices where consecutive pairs are connected by edges. Vertices **can repeat**.
* **Simple Path:** A path where all vertices must be **distinct**.
* **Optimization Logic:** In graphs with non-negative edge weights, the shortest path will naturally be a **simple path**. Any detour involving a cycle only adds cost (or maintains it) without reaching the target faster.


---

## 3. BFS for Unweighted Graphs
When all edge weights are uniform (e.g., all equal to 1), the cost of a path is simply the number of edges. In this specific case, **Breadth-First Search (BFS)** is the optimal algorithm.

### The Layered Progression
BFS explores a graph in "layers."
* **Layer 0:** The source vertex.
* **Layer 1:** Immediate neighbors of the source.
* **Layer $K$:** Vertices reachable in exactly $K$ steps.

### BFS Logic (Pseudocode)
This pseudocode describes how BFS maintains distances and explores the graph using a queue.

```markdown
Procedure BreadthFirstSearch(Graph, Source):
    Initialize DistanceArray with Infinity for all vertices
    Set DistanceArray[Source] = 0
    Initialize Empty Queue
    Add Source to Queue

    While Queue is not empty:
        CurrentVertex = Remove first element from Queue
        
        For each Neighbor of CurrentVertex:
            If DistanceArray[Neighbor] is Infinity:
                DistanceArray[Neighbor] = DistanceArray[CurrentVertex] + 1
                Add Neighbor to Queue
```

### The "Token" Concept for Visualization
To understand layer transitions, imagine placing a **Token** in the queue:
1.  Enqueue the **Source** and a **Token** behind it.
2.  Process all vertices until the Token is reached.
3.  Everything processed before the Token belongs to the current layer.
4.  Once the Token is reached, move it to the back of the queue to mark the end of the next layer.

## 4. Limitations of BFS in Weighted Graphs
BFS fails when edge weights are inconsistent because it assumes the shortest path is always the one with the fewest edges.

* **The Failure Case:** A direct edge from $S$ to $B$ with weight 10 might be "shorter" in steps, but a path $S \rightarrow A \rightarrow B$ with weights $1+1$ is "cheaper" in total cost. BFS would incorrectly choose the direct edge.
* **The "Puncture" Trick:** One could theoretically replace a weighted edge of value $W$ with $W$ edges of value 1 by adding "dummy" vertices. 
    * *Constraint:* This is inefficient and expensive, as the number of vertices would explode based on the size of the weights.

---

## 5. Summary of Algorithm Selection
The lecture categorizes which algorithm to use based on the graph's properties:

| Scenario | Algorithm |
| :--- | :--- |
| **Unweighted Graphs** | Breadth-First Search (BFS) |
| **Positive Weighted Graphs** | Dijkstra's Algorithm |
| **Negative Weights (No Negative Cycles)** | Modified Dijkstra / Bellman-Ford |
| **Negative Weight Cycles** | Bellman-Ford (Detects or Solves) |
| **All-Pairs Shortest Path** | Floyd-Warshall |

### Main Takeaways
* **BFS** is the foundation for shortest paths in unweighted scenarios due to its layered exploration.
* **Negative cycles** disrupt the definition of shortest paths because one could traverse the cycle infinitely to keep reducing the path cost.
* For weighted graphs, we must move beyond BFS to algorithms like **Dijkstra**, which "predicts" the next closest real vertex without needing dummy nodes.
