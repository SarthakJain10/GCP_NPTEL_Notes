# Analysis of Simple Skewness Optimization

This lecture explores an algorithmic problem from the Venture Cup 2016 Elimination Round (found on Codeforces) titled **"Simple Skewness."** It provides a detailed walkthrough on using **sorting** and **binary search** to solve a problem involving statistics and subset selection.

---

## 1. Defining Skewness
In the context of this problem, **skewness** is a mathematical property of a collection of numbers.

* **Definition:** Skewness is the difference between the **mean** and the **median** of a collection of numbers.
* **Formula:** `Skewness = Mean - Median`.
* **Key Characteristics:**
    * Unlike many statistical definitions, this is a literal difference, not an absolute one. It can be positive, negative, or zero.
    * The collection of numbers is a **multi-set**, meaning it can contain repetitions.
    * For an even-sized collection, the median is the average of the two middle elements.


---

## 2. The Problem Statement
Given a collection of integers (which may include duplicates), the goal is to find a **non-empty subset** that yields the **largest possible simple skewness**.

* **Constraints:** The input size ($N$) can be up to 200,000 ($2 \times 10^5$).
* **Implication:** A brute-force approach (checking all subsets) or an $O(N^2)$ solution will be too slow. The target complexity is approximately **$O(N \log N)$**.

---

## 3. Core Observations and Simplifications

### Assuming Odd-Sized Subsets
To simplify the logic, we can assume the optimal subset always has an **odd number of elements**.
* **Logic:** Any subset with zero skewness (like a singleton set) provides a baseline. An optimal solution will never have negative skewness.
* **Reduction:** If an optimal set has an even number of elements, removing the larger of the two middle elements will result in an odd-sized set with skewness that is either the same or higher.

### The Median as an Anchor
If we fix an element to be the median, finding the optimal subset becomes easier. Since the median is fixed, maximizing skewness ($Mean - Median$) is equivalent to **maximizing the mean**.

---

## 4. The Greedy Selection Strategy
When a specific element is chosen as the median and the subset size is fixed to $2k + 1$:
1.  **Selection:** To maximize the mean, we need the largest possible sum.
2.  **Constraint:** To keep the chosen element as the median, we must pick $k$ elements smaller than (or equal to) the median and $k$ elements larger than (or equal to) the median.
3.  **Strategy:** * Pick the **$k$ largest elements** from the very end of the sorted array.
    * Pick the **$k$ largest elements available before the median** (those immediately to its left in a sorted array).

---

## 5. Optimization via Binary Search
Instead of iterating through every possible size $k$ for every possible median (which would be $O(N^2)$), we observe that the mean behaves in a **bitonic** manner .

### Bitonic Behavior of the Mean
As we increase $k$ (adding pairs of elements), the mean tends to increase up to a peak and then decreases.
* **Reasoning:** The pairs of elements we add ($a$ from the left, $b$ from the right) get progressively smaller as we move further from the median and the end of the array.
* **Critical Event:** Once the average of a new pair $(a+b)/2$ is less than or equal to the current mean of the subset, the new mean will decrease and never increase again.

### Binary Search for $k$
We can binary search for the optimal $k$ by checking the "slope":
* If adding more elements (moving from $k$ to $k+1$) increases the mean, move the search range to the **right**.
* If it doesn't increase the mean, move the search range to the **left**.

---

## 6. Algorithmic Structure (Pseudocode)

To efficiently calculate means during the search, the lecture recommends using a **Prefix Sum Array** .

### High-Level Logic
```python
Algorithm: MaximizeSkewness
Input: Array of integers V

1. Sort array V in non-decreasing order.
2. Build a PrefixSum array for V to allow O(1) range sum queries.
3. Initialize MaxSkewness = 0, BestMedianIndex = 0, BestK = 0.

4. For each index i from 0 to N-1 (Guessing i as the median index):
    a. Determine the maximum possible k (L) based on available elements to the left and right.
    b. Binary Search for optimal k in range [0, L]:
        While search range [Low, High] is not empty:
            Mid = (Low + High) / 2
            Calculate current mean with k = Mid
            Calculate next mean by adding the next best pair (k = Mid + 1)
            If NextMean > CurrentMean:
                Low = Mid + 1
            Else:
                High = Mid
    c. Calculate Skewness for the best k found: (CalculatedMean - V[i])
    d. If Skewness > MaxSkewness:
        Update MaxSkewness and store i and k.

5. Output the subset using the stored BestMedianIndex and BestK.
```

## 7. Summary of Takeaways
* **Mean-Median Relationship:** Skewness is defined here as `Mean - Median`.
* **Search and Sort:** Sorting the input allows for a greedy selection of elements to maximize the mean for a fixed median.
* **Bitonic Optimization:** The $O(N^2)$ approach is optimized to **$O(N \log N)$** by recognizing that the mean's growth is bitonic, allowing binary search to find the optimal subset size ($k$) for each candidate median .
* **Implementation Tip:** Prefix sums are essential for maintaining efficiency when repeatedly calculating means of disjoint sub-segments of the array
