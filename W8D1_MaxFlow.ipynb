# Introduction to Network Maximum Flow

This lecture provides an introduction to the **Maximum Flow** (MaxFlow) problem, a fundamental concept in graph theory and network optimization. It covers the intuitive underpinnings of flow networks, formal definitions, and the initial logic for solving these problems.

---

## 1. Concept and Motivation
The MaxFlow problem is not just a specific graph challenge but a **powerful modeling technique**. Many problems that do not appear to be graph-related can be converted into a flow problem and solved using these algorithms.

### The "Truck and Road" Analogy
To understand flow, imagine a city with a **Factory** and a **Shop**:
* **Factory (Source):** Where materials are produced and sent from.
* **Shop (Sink):** The destination where materials are received.
* **Roads (Edges):** One-way paths connecting various locations.
* **Capacity:** Each road has a limit on how many trucks can pass through it simultaneously.
* **Simultaneity:** Trucks are continuously moving; they cannot "wait" at intersections. Whatever enters a location must immediately exit.


---

## 2. Formal Definitions

### Flow Network
A flow network is a **directed graph** $G = (V, E)$ with the following properties:
* **Source ($s$):** A special vertex where flow originates. Typically has an in-degree of 0.
* **Sink/Target ($t$):** A special vertex where flow ends. Typically has an out-degree of 0.
* **Capacity Function ($c$):** Assigns a non-negative number (often an integer) to each edge $(u, v)$, representing the maximum allowable flow.

### Valid Flow
A flow is an assignment of values to edges that must satisfy two primary constraints:
1.  **Capacity Constraint:** For every edge, the flow assigned must be less than or equal to the edge's capacity ($0 \le \text{flow}(u, v) \le c(u, v)$).
2.  **Conservation Constraint:** For every vertex except the source ($s$) and sink ($t$), the total incoming flow must equal the total outgoing flow.
    * *Total Inflow = Total Outflow*

### Value of a Flow
The **value of a flow** is the total amount of material pushed out of the source. Mathematically, this is proven to be identical to the total amount of material accumulated at the sink, as nothing can be "stuck" at intermediate nodes.

---

## 3. The Maximum Flow Problem
The objective is to find a feasible flow that **maximizes** the total value sent from $s$ to $t$.

### Identifying Constraints
The maximum flow is often limited by the "weakest links" in the network.
* **Saturated Edges:** Edges where the current flow equals the capacity (they are "tight" or "used to the brim").
* **Bottleneck:** In any single path, the edge with the minimum capacity determines the maximum flow that can be sent through that specific path.

---

## 4. Greedy Approach (Introductory Logic)
A natural starting point for solving MaxFlow is a greedy strategy. 

### Greedy Logic Steps
1.  Find a path from the source ($s$) to the sink ($t$) where all edges have remaining capacity.
2.  Identify the **bottleneck capacity** (the minimum remaining capacity among all edges in that path).
3.  Push that amount of flow along the path.
4.  Subtract that amount from the capacity of those edges (adjusting the network).
5.  Repeat until no more paths from $s$ to $t$ exist.

### Greedy Pseudocode
```python
function GREEDY_MAX_FLOW(Graph, source, sink):
    Initialize total_flow = 0
    
    while there exists a path from source to sink:
        Find a path P from source to sink using edges with capacity > 0
        
        Identify bottleneck = minimum capacity of any edge in path P
        
        For each edge in path P:
            Decrease capacity of edge by bottleneck
        
        total_flow = total_flow + bottleneck
        
    return total_flow
```

*Note: The lecture concludes by questioning whether this greedy approach is always optimal or if it can get stuck in a "sub-optimal" state. This serves as the bridge to more advanced algorithms like Ford-Fulkerson.*

---

## 5. Summary of Takeaways
* **Flow Networks** consist of a source, a sink, and directed edges with capacities.
* **Flow Conservation** ensures that material is neither created nor destroyed at intermediate nodes.
* The **MaxFlow Value** can be measured either as the total flow exiting the source or entering the sink.
* A **Bottleneck** is the limiting capacity on a specific path that defines how much flow can be added.
* While a **Greedy Strategy** provides a valid flow, further algorithmic refinement is needed to ensure it is the *maximum* possible flow.
