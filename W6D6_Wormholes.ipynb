# Shortest Paths and Negative Cycle Detection: The Wormhole Problem

This lecture focuses on the application of the **Bellman-Ford algorithm** to solve a specific pathfinding challenge known as the "Wormhole Problem." It covers the transition from theoretical shortest path concepts to practical implementation, specifically focusing on graphs with negative edge weights.

---

## 1. Problem Conceptualization: Star Systems and Wormholes
The lecture uses a science-fiction scenario to model a graph theory problem. The goal is to determine if a traveler can move back in time indefinitely by traversing a sequence of wormholes.

### Graph Modeling
To solve the problem, the scenario is translated into a graph abstraction:
* **Vertices**: Represent **star systems**.
* **Edges**: Represent **wormholes** connecting two star systems.
* **Directed Edges**: Wormholes are "one-way only," meaning the graph is directed.
* **Simple Graph**: The problem guarantees no self-loops (wormholes connecting a system to itself) and no multi-edges (at most one wormhole between any pair in a specific direction).
* **Reachability**: It is assumed that every star system is reachable from the starting point (the solar system).

### Edge Weights and Time Travel
The "cost" of traveling through a wormhole is represented by a constant time difference:
* **Positive Weight**: Represents traveling into the **future** (e.g., +15 years).
* **Negative Weight**: Represents traveling into the **past** (e.g., -42 years).
* **The Objective**: Find a **cycle of wormholes** that allows a scientist to end up in the past. In graph terms, this is equivalent to detecting a **negative weight cycle**.

---

## 2. Algorithmic Approach: Bellman-Ford
The Bellman-Ford algorithm is selected because it is designed for the most general edge-weight situations, including those where negative cycles might exist.

### Core Logic
1.  **Initialization**: Set the distance to the source vertex to zero and all other vertices to infinity.
2.  **Relaxation**: Iterate through every edge in the graph and "relax" those that are **tense**. An edge from vertex `U` to `V` with weight `W` is tense if the current distance to `V` is greater than the distance to `U` plus `W`.
3.  **Iteration Count**: The relaxation process is repeated exactly **N - 1** times (where N is the number of vertices).
4.  **Negative Cycle Detection**: After N - 1 iterations, one final pass is made over all edges. If any edge is still tense, it indicates the presence of a negative weight cycle.


### Implementation Strategy
The lecture suggests using an **Edge List** representation for this problem. While adjacency lists or matrices are possible, an edge list is highly efficient for Bellman-Ford because the inner loop simply iterates through all edges sequentially.

* **Performance Note**: Using an adjacency matrix would result in an $O(N^3)$ complexity because scanning all edges requires looking at every cell in the matrix. An edge list or adjacency list keeps the complexity at $O(N \times E)$.

---

## 3. Pseudocode Implementation
The following pseudocode outlines the Bellman-Ford logic as presented in the lecture for detecting a negative cycle.

```python
FUNCTION BellmanFord(SourceVertex, VertexCount, EdgeList):
    # Step 1: Initialization
    Set Distances[all vertices] to INFINITY
    Set Distances[SourceVertex] to 0
    Set Predecessor[all vertices] to NULL

    # Step 2: Relax edges N-1 times
    FOR iteration FROM 1 TO (VertexCount - 1):
        FOR EACH edge (U, V) WITH weight W IN EdgeList:
            IF Distances[U] + W < Distances[V]:
                Distances[V] = Distances[U] + W
                Predecessor[V] = U

    # Step 3: Check for negative cycles
    FOR EACH edge (U, V) WITH weight W IN EdgeList:
        IF Distances[U] + W < Distances[V]:
            # A tense edge still exists; negative cycle detected
            RETURN TRUE (Cycle Exists)

    # No negative cycles found
    RETURN FALSE (No Cycle)
```

## 4. Complexity and Constraints
For the specific "Wormhole Problem" discussed:
* **Number of Star Systems (N)**: Up to 2,000.
* **Number of Wormholes (E)**: Up to 2,000.
* **Total Operations**: Approximately $N \times E = 4 \times 10^6$.
* This volume of operations is well within the time limits for standard computational environments, making Bellman-Ford an efficient choice for these constraints.

## 5. Summary of Takeaways
* **Negative Cycle Detection**: The primary utility of the Bellman-Ford algorithm in this context is its ability to identify negative cycles where the concept of a "shortest path" becomes ill-defined (as the path could infinitely decrease in weight).
* **Modeling Real-World Logic**: Time travel into the past is mathematically modeled as negative edge weights in a directed graph.
* **Implementation Efficiency**: While Bellman-Ford is $O(V \times E)$, using an edge list is the most direct way to implement the inner "relaxation" loop.
* **Transition to All-Pairs**: The lecture concludes by noting that while Bellman-Ford solves the Single-Source Shortest Path (SSSP) problem, more complex problems might require computing paths between all pairs of vertices (leading into the Floyd-Warshall algorithm).
