# Solving the "War" Problem with Disjoint Set Union (DSU)

This lecture provides the solution strategy for the "War" problem, which involves managing friendship and rivalry relationships among $N$ people. The solution leverages the **Disjoint Set Union (DSU)** data structure combined with an additional mechanism to track enemies.

## 1. Data Structures for the Solution
To manage the evolving relationships, two primary structures are used:
* **Standard DSU:** Tracks friendship clusters. People in the same set are friends.
* **Enemy Array:** An array of size $N$ that stores the enemy of each **leader** element. Each leader needs to point to at most one other leader to represent the rivalry between their respective clusters.


## 2. Query Logic
Queries are answered by examining the leader (representative) elements of the sets.

### AreFriends(P, Q)
* **Logic:** Two people are friends if they belong to the same friendship cluster.
* **Mechanism:** Check if `FindRoot(P) == FindRoot(Q)`.

### AreEnemies(P, Q)
* **Logic:** Two people are enemies if the leaders of their respective sets are marked as mutual enemies in the enemy array.
* **Mechanism:** Check if `EnemyOf[FindRoot(P)] == FindRoot(Q)`.

## 3. Handling Operations
The "make" operations are more complex as they must handle inferences and contradictions.

### MakeFriends(P, Q)
This operation attempts to merge the friendship sets of $P$ and $Q$.
1. **Contradiction Check:** If $P$ and $Q$ are already enemies, output `-1`.
2. **Standard Union:** If no contradiction, perform a union of the two sets.
3. **Handle Enemy Inferences:**
    * If $P$'s cluster has an enemy $E_P$ and $Q$'s cluster has an enemy $E_Q$, then $E_P$ and $E_Q$ must become friends (Rule: "Common enemy makes friends").
    * The new leader of the merged $P+Q$ set must point to the new leader of the merged $E_P+E_Q$ set in the enemy array.

### MakeEnemies(P, Q)
This operation establishes a rivalry between the sets of $P$ and $Q$.
1. **Contradiction Check:** If $P$ and $Q$ are already friends, output `-1`.
2. **Handle Friendship Inferences:**
    * If $P$ already has an enemy $E_P$, then $Q$ and $E_P$ must become friends (Rule: "Enemy of my enemy is my friend").
    * Similarly, if $Q$ has an enemy $E_Q$, then $P$ and $E_Q$ must become friends.
3. **Update Enemy Pointers:** Ensure the leaders of the resulting sets point to each other as enemies.


## 4. Pseudocode for Operations

### Main Logic Framework
```python
Function SetFriends(Person_P, Person_Q):
    Root_P = FindRoot(Person_P)
    Root_Q = FindRoot(Person_Q)
    
    If Root_P == Root_Q:
        Return # Already friends
    
    If AreEnemies(Root_P, Root_Q):
        Output -1
        Return

    # Identify existing enemies before merging
    Enemy_P = EnemyArray[Root_P]
    Enemy_Q = EnemyArray[Root_Q]

    # Merge P and Q clusters
    NewRoot_PQ = Union(Root_P, Root_Q)

    # Apply Common Enemy Rule
    If Enemy_P exists and Enemy_Q exists:
        NewRoot_Enemy = Union(Enemy_P, Enemy_Q)
        UpdateEnemyPointers(NewRoot_PQ, NewRoot_Enemy)
    Else If Enemy_P exists:
        UpdateEnemyPointers(NewRoot_PQ, Enemy_P)
    Else If Enemy_Q exists:
        UpdateEnemyPointers(NewRoot_PQ, Enemy_Q)
```

## 5. Summary of Takeaways
* **Leader-to-Leader Rivalry:** Tracking enemies only at the leader level is sufficient to represent all group-wide rivalries.
* **Recursive Merging:** An operation like `MakeFriends` can trigger a secondary `MakeFriends` (via the `Union` of enemies) to maintain consistency with the three rules.
* **Consistency:** The system maintains a state where clusters pair up as enemies or remain isolated, ensuring no leader ever has more than one enemy leader.
