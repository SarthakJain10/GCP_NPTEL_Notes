# Dynamic Programming: Top-Down Approach with the Frog Problem

This lecture introduces the concept of **Dynamic Programming (DP)** through a classic problem called "Frog 1." It focuses on the **top-down approach**, also known as **memoization**, to optimize recursive solutions.

---

## 1. Problem Overview: Frog 1
The problem involves a frog navigating a sequence of stones. The goal is to reach the final stone with the **minimum total cost**.

### The Setup
* **Stones:** There are $N$ stones, numbered $1$ to $N$.
* **Heights:** Each stone $i$ has a specific height $H_i$.
* **The Frog:** Starts at stone $1$ and aims to reach stone $N$.

### The Rules of Movement
When the frog is at stone $i$, it can make one of two types of jumps:
1. **Small Jump:** To stone $i + 1$.
2. **Large Jump:** To stone $i + 2$.

### The Cost Calculation
Every jump carries a cost equal to the **absolute difference** between the heights of the two stones:
* **Cost of jump from $i$ to $j$:** $|H_i - H_j|$
* **Total Cost:** The sum of costs of all individual jumps in the sequence.


---

## 2. Theoretical Framework for the Solution
To solve this, the lecture explores why simple strategies fail and how a recursive structure emerges.

### Why Greedy Fails
A **greedy strategy** (always picking the cheapest immediate jump) does not work because a cheap jump now might lead to a very expensive set of jumps later. 

### Why Brute Force Fails
Enumerating every possible sequence of jumps results in an **exponential number of paths**. For $N = 10^5$ (as seen in competitive programming constraints), this is computationally impossible.

### The Recursive Intuition (Working Backward)
Instead of looking forward from stone $1$, consider how the frog arrives at the last stone ($N$):
* To reach stone $N$, the frog must have come from either **stone $N-1$** or **stone $N-2$**.
* Therefore, the optimal cost to reach $N$ is the minimum of:
    1. Optimal cost to reach $N-1$ + Cost of jump $(N-1 \to N)$.
    2. Optimal cost to reach $N-2$ + Cost of jump $(N-2 \to N)$.

---

## 3. Algorithm Design
The logic can be formalized into a recursive function and base cases.

### Base Cases
Base cases are essential to stop the recursion and handle the start of the journey:
* **One Stone ($N=1$):** The frog is already at the destination. Cost = **0**.
* **Two Stones ($N=2$):** Only one jump is possible (from $1$ to $2$). Cost = **$|H_1 - H_2|$**.

### Pseudocode Representation
The following pseudocode represents the logic of finding the minimum cost recursively:

```markdown
FUNCTION GetMinCost(n, heights)
    // Base Case: Already at the first stone
    IF n equals 1 THEN
        RETURN 0
    
    // Base Case: Only one way to reach the second stone
    IF n equals 2 THEN
        RETURN AbsoluteDifference(heights[1], heights[2])
    
    // Recursive Step: Option A (coming from n-1)
    costA = GetMinCost(n - 1, heights) + AbsoluteDifference(heights[n], heights[n-1])
    
    // Recursive Step: Option B (coming from n-2)
    costB = GetMinCost(n - 2, heights) + AbsoluteDifference(heights[n], heights[n-2])
    
    // Return the better of the two options
    RETURN Minimum(costA, costB)
END FUNCTION
```

---

## 4. Complexity Analysis
Analyzing the performance of the raw recursive approach:

* **Recursion Tree:** Every call to the function (for $N$) triggers two more calls ($N-1$ and $N-2$). This creates a tree structure.
* **Redundancy:** The tree reveals that the same values (e.g., `GetMinCost(3)`) are calculated multiple times across different branches.
* **Time Complexity:** The number of nodes in this tree grows similarly to the **Fibonacci sequence**, resulting in **exponential time complexity** ($O(2^n)$). This makes the solution too slow for large inputs.


---

## Summary of Takeaways
* **Dynamic Programming** is a technique used to solve complex problems by breaking them into simpler subproblems.
* The **Recursive Formula** for the Frog problem relies on the fact that the state at stone $N$ depends only on the states of stones $N-1$ and $N-2$.
* **Base Cases** (for stones 1 and 2) are critical to ensure the algorithm terminates and provides a starting point for calculations.
* A **Naive Recursive Solution** is inefficient due to **overlapping subproblems** (redundant work), leading to exponential growth in computation time.
