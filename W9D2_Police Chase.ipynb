# Solving the Police Chase Problem using Minimum Cut

This lecture explores the application of the **Max-Flow Min-Cut Theorem** to solve the "Police Chase" problem from the CSES problem set. The core focus is on how to transition from finding a maximum flow to identifying the specific edges that constitute a minimum cut.

## 1. Problem Overview: The Police Chase
The problem presents a scenario where a bank robber is moving from a bank to a harbor. The goal for the police is to block the minimum number of streets to ensure there is no possible path between these two locations.

### Key Components:
* **Source ($S$):** The location of the bank (represented as vertex 1).
* **Target ($T$):** The location of the harbor (represented as vertex $n$).
* **Streets:** Undirected connections between locations.
* **Objective:** Find the smallest set of edges (streets) to remove to disconnect $S$ from $T$.


## 2. Modeling with Flow Networks
To solve this, the city's street map is modeled as a flow network:
* Each street is treated as an edge with a **capacity of 1**. 
* Because the goal is to minimize the *number* of streets, assigning a uniform capacity of 1 ensures that the minimum cut capacity corresponds directly to the minimum number of edges.
* The problem is a direct application of the **Minimum Cut**, which, according to the Max-Flow Min-Cut Theorem, is equal in value to the **Maximum Flow**.

## 3. Algorithm Strategy
The solution follows a three-step process:
1.  **Compute Maximum Flow:** Run an algorithm like **Ford-Fulkerson** (specifically the **Edmonds-Karp** implementation) to find the maximum flow from the bank to the harbor.
2.  **Identify Reachable Vertices:** After the flow is maximized, look at the **residual graph**. Find all vertices that are still reachable from the source vertex using edges with remaining capacity.
3.  **Extract the Cut Edges:** Identify edges where one endpoint is in the "reachable" set and the other endpoint is in the "unreachable" set.

## 4. Implementation Logic

### Identifying the Reachable Set
To find the set of vertices reachable from the source in the final residual graph, a modified **Breadth-First Search (BFS)** is used.

```pseudocode
Function GetReachableSet(SourceVertex, ResidualGraph):
    Create an empty list called ReachableSet
    Create a Queue and add SourceVertex to it
    Mark SourceVertex as visited
    Add SourceVertex to ReachableSet

    While Queue is not empty:
        Current = Remove front of Queue
        For each Neighbor of Current in ResidualGraph:
            If Neighbor is not visited AND Capacity(Current, Neighbor) > 0:
                Mark Neighbor as visited
                Add Neighbor to ReachableSet
                Add Neighbor to Queue
    
    Return ReachableSet
```

### Finding the Minimum Cut Edges
Once you have the set of vertices reachable from the source ($S_{comp}$) and those that are not ($T_{comp}$), the edges crossing between these two sets form the minimum cut.

```pseudocode
Algorithm FindMinCutEdges:
    1. Run Max-Flow algorithm to update the ResidualGraph.
    2. S_comp = GetReachableSet(Source, ResidualGraph)
    
    3. Define T_comp as all vertices in the graph NOT in S_comp.
    
    4. For each vertex U in S_comp:
        For each vertex V in T_comp:
            If there is an original street between U and V:
                Print or Store Edge(U, V) as part of the Minimum Cut
```

## 5. Important Considerations
* **Output Requirements:** If the problem only asks for the *number* of streets, the Max-Flow value is sufficient. However, to list the actual streets, you must perform the BFS on the residual graph as described.
* **Residual Graph:** The "reachable" check must be performed on the residual graph after the max-flow algorithm has finished, as this graph indicates where additional flow *could* have gone.
* **Edge Directions:** While the streets are undirected, in flow networks they are often modeled as bidirectional edges with capacity. The cut edges are those that connect the source-connected component to the rest of the graph.

## Summary of Takeaways
* The **Police Chase** problem is a classic application of finding a **Minimum Cut** in a graph where all edge capacities are set to 1.
* The **Max-Flow** value gives the minimum number of edges to remove, but a **BFS on the residual graph** is required to identify which specific edges to remove.
* Vertices reachable from the source in the final residual graph form one side of the cut; any edge connecting a vertex in this set to a vertex outside of it is a bottleneck edge contributing to the cut.
