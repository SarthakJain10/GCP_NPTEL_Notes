# Ford-Fulkerson Algorithm for Maximum Flow

This lecture covers the **Ford-Fulkerson algorithm**, a fundamental method for finding the maximum flow in a network. It addresses the limitations of simpler greedy approaches and introduces the critical concept of **residual graphs** and **back edges**.

---

## 1. Limitations of the Simple Greedy Approach
A basic approach to finding maximum flow might involve repeatedly finding a path from the source ($s$) to the sink ($d$) and pushing as much flow as possible. However, this method can fail because:
* **Irreversibility:** Once an edge is saturated (flow equals capacity), it effectively disappears from the graph in a simple model.
* **Suboptimal Choices:** Selecting a specific path early on can "block" other paths that would have contributed to a higher total flow.
* **Example:** In a graph where a central edge is shared by two potential paths, using that central edge fully might result in a flow of 1, whereas avoiding it could have allowed two separate paths to provide a total flow of 2.


---

## 2. The Residual Graph and Undo Mechanism
To fix the mistakes of a greedy approach, the algorithm needs a way to "undo" or reroute flow. This is achieved using a **Residual Graph** ($G_f$).

### Key Components of the Residual Graph:
* **Forward Edges:** If an edge $(a, b)$ in the original network has capacity $C$ and current flow $f$, the residual graph contains an edge $(a, b)$ with a **residual capacity** of $C - f$. This represents how much more flow can be sent.
* **Back Edges:** For every original edge $(a, b)$ with flow $f$, an artificial edge $(b, a)$ is added to the residual graph with capacity $f$. This represents the ability to "push back" or cancel existing flow.

### The Role of Back Edges:
When the algorithm finds a path in the residual graph that uses a back edge, it is effectively **rolling back** a previous decision. Instead of physically moving flow backward, it reduces the flow on the corresponding forward edge in the original network.

---

## 3. The Ford-Fulkerson Algorithm
The algorithm iteratively improves the flow until no more augmenting paths can be found in the residual graph.

### High-Level Logic
```
function FordFulkerson(Graph, source, sink):
    Initialize flow on all edges to 0
    Initialize MaxFlow to 0
    
    While there exists a path P from source to sink in the Residual Graph:
        1. Find the Bottleneck Capacity (F):
           F = minimum residual capacity of all edges in path P
        
        2. Augment the flow:
           For each edge (u, v) in path P:
               If (u, v) is a Forward Edge:
                   Increase flow of (u, v) in original graph by F
               Else if (u, v) is a Back Edge:
                   Decrease flow of original edge (v, u) by F
        
        3. Update MaxFlow:
           MaxFlow = MaxFlow + F
           
    Return MaxFlow
```

---

## 4. Implementation Details and Optimizations

### Tracking Flow vs. Capacity
In practice, instead of constantly rebuilding the graph, developers often:
* Keep capacities fixed.
* Maintain a **flow function** that can take negative values for back edges.
* The residual capacity of an edge $(u, v)$ is calculated as `Capacity(u, v) - Flow(u, v)`.
* When flow $F$ is pushed through $(u, v)$, the flow on $(u, v)$ increases by $F$, and the flow on the reverse edge $(v, u)$ decreases by $F$.

### Running Time and Efficiency
* **Basic Ford-Fulkerson:** The complexity is $O(m \times \text{MaxFlow})$, where $m$ is the number of edges. This is inefficient if the flow values are very large.
* **Edmonds-Karp Algorithm:** An improvement that uses **Breadth-First Search (BFS)** to always pick the shortest path (in terms of number of edges) in the residual graph. This guarantees a complexity of $O(n \times m^2)$, where $n$ is the number of vertices.
* **Dinic's Algorithm:** A more advanced optimization commonly used in competitive programming. It uses "blocking flows" to reduce the number of iterations to $O(n)$, resulting in an overall complexity of $O(n^2 m)$.

---

## Summary of Key Takeaways
* **Greedy is not enough:** Simple path-finding doesn't work for Max Flow because it cannot undo bad decisions.
* **Residual Graphs are essential:** They provide a formal mechanism for both increasing flow (forward edges) and canceling flow (back edges).
* **Augmenting Paths:** The core of the algorithm is finding any path from source to sink in the residual graph.
* **Algorithm Choice Matters:** While Ford-Fulkerson defines the framework, using BFS (Edmonds-Karp) or Dinic's optimization is necessary for handling large-scale networks efficiently.
