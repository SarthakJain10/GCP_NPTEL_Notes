# Stable Marriage Problem: Concepts and Algorithms

This lecture introduces a classic computer science problem known as the **Stable Marriage Problem**, exploring its mathematical foundations, the concept of stability, and a greedy-based algorithm to solve it.

---

## 1. Problem Overview
The stable marriage problem is a fundamental matching problem involving two disjoint sets of equal size.

### The Setting
* **Participants**: Two groups of people, traditionally labeled as **Men** ($n$) and **Women** ($n$), for a total of $2n$ individuals.
* **Preferences**: Every individual has a **ranking** (a permutation) of all members of the opposite group. These rankings are strictly personal and individual.
* **Goal**: Find a **bijection** (a one-to-one matching) between men and women such that the matching is **stable**.


---

## 2. Defining Stability
To understand a stable matching, we must first define what makes a matching unstable.

### The Blocking Pair
A **blocking pair** occurs in a given matching when a man ($M$) and a woman ($W$) are not matched to each other, but they **both prefer each other** over their current partners.
* If $M$ is matched with $W'$, but prefers $W$.
* If $W$ is matched with $M'$, but prefers $M$.
* $M$ and $W$ have a mutual incentive to break their current alliances to be with each other.

### Stable Matching
A matching is considered **stable** if it contains **no blocking pairs**. Every participant is either with their preferred choice or with someone they prefer over any other person who would actually want to be with them.

---

## 3. The Gale-Shapley Algorithm
This algorithm (often called the Propose-and-Reject algorithm) finds a stable matching in any instance of the problem.

### Key Logic
* **Men Propose**: Men go down their preference list from best to worst.
* **Women Filter**: Women can accept a proposal tentatively or "trade up" if they receive a better offer later.
* **Stability over Loyalty**: If a woman is already matched but receives a proposal from a man she prefers more, she **must** break the current engagement to accept the new one to prevent a blocking pair from forming later.

### Pseudocode Representation
```python
INITIALIZE all men and women as FREE
WHILE there exists a Man (M) who is FREE and has not proposed to every Woman:
    Choose the highest-ranked Woman (W) on M's list to whom he hasn't proposed
    
    IF W is FREE:
        Match (M, W) as an engagement
    ELSE IF W is already matched with Man (M_prime):
        IF W prefers M over M_prime:
            Match (M, W)
            Set M_prime to FREE
        ELSE:
            W rejects M (M remains FREE)
            
RETURN the set of all final engagements
```

---

## 4. Analysis and Proofs

### Termination
The algorithm is guaranteed to end. 
* There are $n$ men and $n$ women.
* Each man proposes to a woman at most once.
* The maximum number of proposals is $n^2$. Since the number of proposals is finite, the loop must terminate.

### Completeness
By the end of the algorithm, everyone is matched. 
* Once a woman receives her first proposal, she is never "free" again; she only switches to better partners.
* If a man remains unmatched, it would imply there is a woman he hasn't proposed to who is also unmatched. However, the algorithm only stops when no free man has anyone left to propose to.

### Stability Proof (by Contradiction)
Suppose there is a blocking pair $(M, W)$ in the final matching where $M$ is matched with $W'$ and $W$ is matched with $M'$.
1.  Since $M$ prefers $W$ over $W'$, he must have proposed to $W$ **before** he proposed to $W'$.
2.  If $M$ proposed to $W$, $W$ must have either rejected him immediately or dumped him later for someone she liked better.
3.  Because women only "trade up," the fact that $W$ is not with $M$ means her final partner ($M'$) must be someone she likes **more** than $M$.
4.  This contradicts the definition of a blocking pair, proving the final matching is stable.

---

## 5. Summary of Main Takeaways
* **Stable Matching Always Exists**: For any set of preferences, there is at least one stable matching.
* **Mutual Incentive**: Stability is not about making everyone perfectly happy, but about ensuring no two people have a mutual incentive to "cheat" on the system.
* **Algorithm Efficiency**: The Gale-Shapley algorithm is efficient, completing in $O(n^2)$ time relative to the number of participants.
* **Mechanism**: Men improve their situation by moving down their lists (starting from best), while women improve their situation by "trading up" (accepting better incoming proposals).
