# Breadth-First Search (BFS) and Depth-First Search (DFS) Lecture Notes

This lecture covers the fundamental concepts, implementations, and properties of two primary graph traversal algorithms: **Breadth-First Search (BFS)** and **Depth-First Search (DFS)**.

---

## 1. Introduction to Graph Traversals
Graph traversal is the process of visiting every vertex (node) in a graph. The lecture builds upon previous knowledge of graph representations, specifically **Adjacency Matrices** and **Adjacency Lists**.


## 2. Breadth-First Search (BFS)
BFS explores the graph in "layers." It starts at a source vertex and visits all its immediate neighbors before moving to the neighbors' neighbors.

### 2.1 The Wikipedia Analogy
Think of BFS like browsing Wikipedia:
1. You start on a main page.
2. You open all the links on that page in **background tabs**.
3. You finish the current page, then move to the first background tab you opened and repeat the process.

### 2.2 Key Concepts in BFS
* **Queue-based:** BFS uses a **Queue** (First-In-First-Out) data structure to manage the order of exploration.
* **Distance:** It tracks the distance of each vertex from the source. The distance is defined as the number of edges in the shortest path.
* **Layers:** Vertices are visited in order of their distance from the source (Layer 0, then Layer 1, etc.).
* **Parent Pointers:** Each vertex (except the source) stores a reference to the vertex that first "discovered" it.

### 2.3 BFS Pseudocode Implementation
The following pseudocode outlines the BFS logic using an adjacency list and a queue.

```markdown
Procedure BFS(Graph, SourceVertex):
    Initialize Distance array with Infinity for all vertices
    Initialize Parent array with Null for all vertices
    
    Set Distance[SourceVertex] = 0
    Initialize Empty Queue
    Push SourceVertex onto Queue
    
    While Queue is not empty:
        CurrentVertex = Remove from front of Queue
        
        For each Neighbor of CurrentVertex:
            If Distance[Neighbor] == Infinity: // Neighbor is unvisited
                Set Distance[Neighbor] = Distance[CurrentVertex] + 1
                Set Parent[Neighbor] = CurrentVertex
                Push Neighbor to back of Queue
```

## 3. Depth-First Search (DFS)
DFS explores as far as possible along a branch before backtracking. It follows a single path until it hits a "dead end."

### 3.1 The Wikipedia Analogy
Think of DFS like a different browsing style:
1. You click the first link you see.
2. On the new page, you immediately click the first link *that* page has.
3. You only go back to previous pages (backtrack) when you reach a page with no new links.

### 3.2 Key Concepts in DFS
* **Recursion/Stack-based:** DFS is most naturally implemented using **recursion**, which implicitly uses a system stack.
* **Backtracking:** When a vertex has no more unvisited neighbors, the algorithm "backtracks" to the previous vertex to continue exploration.
* **Visited Status:** Crucial to mark vertices as "visited" immediately upon entry to avoid infinite loops.

### 3.3 DFS Pseudocode Implementation (Recursive)
This implementation uses a global or passed-in array to track visited status.

```markdown
Procedure DFS(CurrentVertex):
    Mark CurrentVertex as VISITED
    // (Optional: Perform actions like printing the vertex here)
    
    For each Neighbor of CurrentVertex:
        If Neighbor is UNVISITED:
            DFS(Neighbor) // Recursive call
            
    // Control returns (backtracks) once all neighbors are explored
```

## 4. Complexity and Performance
The efficiency of these traversals depends heavily on how the graph is represented.

| Representation | Complexity | Notes |
| :--- | :--- | :--- |
| **Adjacency Matrix** | $O(V^2)$ | Inefficient for sparse graphs; checking neighbors takes $O(V)$ for every vertex. |
| **Adjacency List** | $O(V + E)$ | Superior for most cases, especially sparse graphs. $V$ = Vertices, $E$ = Edges. |

*Note: In Competitive Programming, the Adjacency List is preferred when the number of vertices exceeds a few thousand due to memory and time constraints.*

## 5. Properties and Applications

### 5.1 BFS Properties
* **Shortest Path:** BFS finds the shortest path (in terms of edge count) from the source to any reachable vertex in an unweighted graph.
* **No Level Jumps:** In a BFS tree, edges only exist between consecutive levels or within the same level. You cannot have an edge skipping from Level 2 to Level 5.

### 5.2 DFS Properties
* **Back Edges:** In undirected graphs, DFS only results in "tree edges" (edges used for discovery) and "back edges" (edges pointing to an ancestor in the tree). There are no "cross edges" between different branches.
* **Leaf Property:** In the final state of a DFS tree, there are no edges between the leaves (terminal nodes).

### 5.3 Common Applications
Both algorithms are foundational for:
* **Connected Components:** Finding groups of connected nodes.
* **Bipartiteness:** Checking if a graph can be colored with two colors.
* **Topological Sort:** Ordering nodes (useful for scheduling tasks).
* **Cycle Detection:** Determining if a graph contains a loop.

---

## Summary of Main Takeaways
* **BFS** is "wide" exploration using a **Queue** and is ideal for finding **shortest paths** in unweighted graphs.
* **DFS** is "deep" exploration using **Recursion** and is useful for exploring all possible paths and complex connectivity properties.
* **Adjacency Lists** are generally the most efficient way to store graphs for these algorithms, providing $O(V+E)$ complexity.
* **Visited marking** is the most critical step in both algorithms to prevent infinite loops and redundant processing.
