# Implementing Dijkstra's Algorithm: The "Sending Email" Problem

This lecture focuses on solving the Single Source Shortest Path (SSSP) problem using Dijkstra's Algorithm, specifically within the context of the "Sending Email" problem from the UVA platform. It covers data structure choices, complexity analysis, and implementation details for undirected graphs with non-negative edge weights.

## 1. Problem Context: Sending Email
The lecture uses the "Sending Email" problem to demonstrate a practical application of Dijkstra's algorithm.

### Problem Formulation
* **Entities**: $N$ SMTP servers and $M$ network cables.
* **Graph Model**: Servers are represented as **vertices**, and cables are represented as **edges**.
* **Edge Weights**: Each cable has a **latency** (measured in milliseconds), which acts as the weight of the edge.
* **Goal**: Find the shortest time (minimum weight path) to send a message from a source server $S$ to a target server $T$.
* **Key Constraints**:
    * No delay occurs at the servers themselves.
    * Edge weights are between 0 and 10,000 (non-negative).
    * Cables are **bi-directional** (the graph is undirected).


---

## 2. Algorithm Choice and Data Structures

### Why Dijkstra's?
Since the edge weights are guaranteed to be **non-negative**, Dijkstra’s algorithm is suitable and guaranteed to be correct. For unweighted graphs (all weights = 1), Breadth-First Search (BFS) would suffice, but Dijkstra handles arbitrary non-negative weights.

### Data Structure Challenges
The standard Dijkstra algorithm requires two primary operations:
1.  **Extract Min**: Find and remove the vertex with the smallest current distance.
2.  **Decrease Key**: Update the distance of a vertex already in the collection.

**Implementation Nuances:**
* **Priority Queue (C++ STL/Python)**: Often lacks a direct `decrease_key` method.
* **Set (C++)**: Internally based on balanced Binary Search Trees (BSTs). It allows for insertion, removal, and search in $O(\log N)$ time. To "decrease a key," you delete the old value and insert the updated one.
* **Python Considerations**: Python's built-in priority queue (heapq) also lacks `decrease_key`. Users can either mark old entries as "removed" and insert new ones or use a modified version of Dijkstra that avoids explicit updates.

---

## 3. Complexity Analysis
The performance of Dijkstra’s algorithm depends on the number of vertices ($N$) and edges ($M$).

* **Operations**:
    * Every vertex is extracted exactly once: $N \times \text{Extract Min}$.
    * Every edge is relaxed at most once: $M \times \text{Decrease Key}$.
* **Total Cost**: Using a heap or set, both operations cost $O(\log N)$.
* **Simplified Complexity**: $O((M + N) \log N)$, which is typically expressed as **$O(M \log N)$** since $M$ usually dominates $N$ in most graphs.

## 4. Pseudocode Implementation

### Initialization
Before the main loop, we must set up the distances and the priority collection.

```python
Function Initialize(Graph, Source_Vertex):
    Create a Distance array of size N
    For each vertex v in Graph:
        Distance[v] = INFINITY
    
    Distance[Source_Vertex] = 0
    
    Create an empty Priority_Set (stores pairs of [distance, vertex])
    Insert [0, Source_Vertex] into Priority_Set
    For all other vertices v:
        Insert [INFINITY, v] into Priority_Set
```

### Main Algorithm Body
The core logic involves repeatedly picking the "closest" vertex and relaxing its neighbors.

```python
Function Dijkstra(Graph, Source, Target):
    While Priority_Set is not empty:
        # Extract Min
        Current_Pair = Get_Smallest_Element(Priority_Set)
        Current_Dist = Current_Pair.distance
        U = Current_Pair.vertex
        Remove Current_Pair from Priority_Set
        
        # Edge Relaxation
        For each Neighbor V of U with weight W:
            If (Distance[U] + W) < Distance[V]:
                # Simulate Decrease Key
                Remove [Distance[V], V] from Priority_Set
                Distance[V] = Distance[U] + W
                Insert [Distance[V], V] into Priority_Set
                
                # Optional: Store path info
                Predecessor[V] = U

    Return Distance[Target]
```

## 5. Path Reconstruction
If the problem requires printing the actual path (not just the weight), a **Predecessor Array** is used.

**Logic for Printing Path:**
1.  Start at the target vertex $T$.
2.  Follow the pointers in the `Predecessor` array back to the source $S$.
3.  Store these vertices in a list.
4.  **Reverse** the list to show the path from $S$ to $T$.

## 6. Summary of Key Takeaways
* **Undirected Graphs**: Dijkstra works perfectly on undirected graphs as long as you represent each cable as two directed edges (one in each direction).
* **Non-negative Weights**: This is a strict requirement for the original Dijkstra algorithm to ensure correctness.
* **Efficiency**: The $O(M \log N)$ complexity makes it highly efficient for graphs with tens of thousands of nodes/edges.
* **Implementation Trick**: Using a **Set** (in C++) or **Mark-and-Insert** (in Python) effectively overcomes the lack of a `decrease_key` function in standard libraries.
