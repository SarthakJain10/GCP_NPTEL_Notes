# All-Pairs Shortest Paths: Page Hopping and Floyd-Warshall Implementation

This lecture focuses on the implementation of the **Floyd-Warshall algorithm** to solve the **All-Pairs Shortest Paths (APSP)** problem. Using the "Page Hopping" problem (ICPC World Finals 2000) as a case study, the lecture explores the logic, implementation details, and common pitfalls like integer overflow.

---

## 1. Problem Overview: Page Hopping
The "Page Hopping" problem serves as a practical application for finding shortest paths between all possible pairs of nodes in a network.

### Key Characteristics
* **Graph Type**: Directed graph representing web pages (nodes) and links (edges).
* **Constraints**: Small graph size with a maximum of **100 vertices**.
* **Connectivity**: The problem guarantees **strong connectivity**, meaning a path exists between every pair of nodes. This ensures no infinite distances in the final output.
* **Goal**: Calculate the **average shortest path length** between all arbitrary pairs of nodes.

---

## 2. Implementation Strategy: Floyd-Warshall Algorithm
The Floyd-Warshall algorithm is preferred here due to the small vertex count ($N \le 100$) and the simplicity of its implementation. While running a Breadth-First Search (BFS) $N$ times is a valid alternative for unweighted graphs, Floyd-Warshall is more concise.

### Data Structures
* **Set for Vertices**: Used during input reading to track unique vertex labels, as the total number of vertices ($N$) isn't explicitly provided.
* **Distance Matrix**: A 2D array (e.g., `dist[100][100]`) initialized to store the shortest known distance between any two nodes $i$ and $j$.


### Initialization
1.  **Self-Loops**: Distance from a vertex to itself is set to $0$ (`dist[i][i] = 0`).
2.  **Edges**: For every direct link from $A$ to $B$, initialize `dist[A][B] = 1`.
3.  **Non-edges**: All other pairs are initialized to a value representing **Infinity**.

### Algorithm Logic (Pseudocode)
The core of the algorithm consists of three nested loops. The outer loop iterates through "intermediate" vertices, while the inner loops update distances between all pairs $(i, j)$ by checking if a path through the intermediate vertex $k$ is shorter.

```markdown
FOR each intermediate vertex k from 1 to N:
    FOR each starting vertex i from 1 to N:
        FOR each ending vertex j from 1 to N:
            IF distance(i, k) + distance(k, j) < distance(i, j):
                SET distance(i, j) = distance(i, k) + distance(k, j)
```

---

## 3. Handling "Infinity" and Preventing Overflows
A critical detail in shortest path algorithms is how to represent and handle the value for **Infinity**.

### The "Large Enough" Rule
The value for Infinity must be greater than the maximum possible path weight in the graph. A safe bound is:
$$\text{Infinity} > (\text{Total number of edges}) \times (\text{Maximum edge weight})$$

### The Overflow Trap
Using the maximum possible integer value (e.g., `INT_MAX`) can cause **integer overflow**:
* If `dist[i][k]` is Infinity and you add an edge weight $w$ to it, the value may wrap around to a large negative number.
* The algorithm will incorrectly identify this negative value as a "shorter" path than the current `dist[i][j]`.
* **Solution**: Set Infinity to a sufficiently large value that is still small enough to allow additions without exceeding the data type's limit, or use larger data types like `long`.

---

## 4. Complexity and Suitability
* **Time Complexity**: $O(N^3)$, where $N$ is the number of vertices.
* **Practical Limit**: Typically suitable for graphs where $N \le 450$. Beyond this, the cubic growth makes the algorithm too slow for standard time limits.
* **Advantage**: Extremely short code (approximately 4 lines for the core logic), making it ideal for competitive programming when constraints allow.

---

## Summary of Main Takeaways
* **Floyd-Warshall** is an elegant $O(N^3)$ solution for the **All-Pairs Shortest Path** problem, especially effective for small, dense graphs.
* The algorithm works by iteratively refining the shortest path between all pairs by considering every vertex as a potential intermediate point.
* Properly initializing the **Distance Matrix** and choosing a safe **Infinity** value (to avoid overflow) are essential for a bug-free implementation.
* For unweighted graphs like the Page Hopping problem, Floyd-Warshall provides the same results as running BFS $N$ times but with a simpler coding structure.
