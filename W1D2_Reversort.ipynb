## Problem Overview
**Reversort** sorts a list of distinct integers in ascending order using reverse operations on carefully chosen subarrays.  
The algorithm iterates through the array, finding the minimum in the unsorted suffix and reversing the subarray from the current position to that minimum's location.  
No early termination occurs even if sorted early; it completes n-1 iterations, with the last element falling into place automatically.  

## Algorithm Pseudocode
```
REVERSE_SORT(L):
    n = length(L)
    for i = 1 to n-1:
        min_idx = POSITION_OF_MINIMUM(L[i..n])
        REVERSE(L[i..min_idx])
```
- Loop runs from i=1 to n-1 (0-based: i=0 to n-2).  
- Finds minimum in subarray L[i..n], reverses from i to its position.  
- Ensures ith smallest element reaches position i each iteration.

## Cost of Reversort
Cost of each reverse is subarray length (j - i + 1, where i to j reversed).  
Total cost: sum over all iterations.  
- Example: costs 3 + 1 + 2 = **6**.  
- Example: costs 7 + 1+1+1+1+1 = **12** (last iteration free). 

## Solution Approach
Simulate algorithm: track cost while finding min position and reversing subarrays.  
Maximum cost occurs when each iteration reverses full remaining subarray (e.g., reverse-sorted arrays yield high costs like 14 for n=5).

- Compute cost:

  ```cpp
  int cost = 0;
  for(int i = 0; i < n-1; i++) {
      auto min_it = min_element(L.begin() + i, L.end());
      auto x = find(L.begin(), L.end(), *min_it) - L.begin();
      reverse(L.begin() + i, L.begin() + x + 1);
      cost += (x - i + 1);
  }
  ```
