## 1. Fundamental Graph Concepts

Before solving the specific problem, the lecture establishes several key definitions essential for understanding graph theory.

### 1.1. Trees
In graph theory, a **tree** is defined as a graph that is both **connected** and **acyclic**.
* **Connected**: A graph is connected if every vertex is reachable from every other vertex via a path. A path is a sequence of vertices where consecutive vertices are joined by a direct edge.
* **Acyclic**: The graph contains no cycles. A **cycle** is a path where the start and end vertices are connected by a direct edge.


### 1.2. Bipartite Graphs
A **bipartite graph** is a graph whose set of vertices can be partitioned into two independent sets (often called "parts") such that every edge in the graph connects a vertex in one part to a vertex in the other.
* No edges exist within the same part.
* A **complete bipartite graph** occurs when every possible edge between the two parts exists.
* **Maximum Edges**: If the two parts have $a$ and $b$ vertices respectively, the maximum number of edges in a bipartite graph is the product $a \times b$.


## 2. Properties of Trees and Bipartiteness

A central premise of the lecture is that **every tree is a bipartite graph**. 

### 2.1. Why Trees are Bipartite
The lecture provides three methods/hints to prove this:
1.  **Induction**: Using the fact that every tree with at least two vertices has at least one **leaf** (a vertex with exactly one neighbor).
2.  **Odd Cycle Characterization**: A graph is bipartite if and only if it has no **odd cycles**. Since trees have no cycles at all, they satisfy this condition.
3.  **Distance-based Partitioning**: Choosing an arbitrary root vertex and partitioning based on distance:
    * Vertices at **even distances** from the root go to one part.
    * Vertices at **odd distances** from the root go to the other part.

### 2.2. Uniqueness of Bipartition
For a connected graph like a tree, the bipartite partition is **unique** (up to swapping the two parts). This means once you assign the first vertex to a side, the placement of all other reachable vertices is forced.

## 3. Problem Analysis: Mahmoud and Ehab and the Bipartiteness

**Objective**: Given a tree with $n$ vertices, add the maximum number of edges possible such that the graph remains both **simple** (no multiple edges between the same two vertices) and **bipartite**.

### 3.1. The Solution Logic
1.  Identify the unique bipartite partition of the given tree into two sets, $A$ and $B$.
2.  Calculate the sizes of these sets: $a = |A|$ and $b = |B|$.
3.  The maximum edges a bipartite graph with these parts can have is $a \times b$.
4.  The tree already contains $n-1$ edges.
5.  **Final Calculation**: The number of edges to add is $(a \times b) - (n - 1)$.

## 4. Implementation Using Traversals

To find the sizes of parts $A$ and $B$, we can use either BFS or DFS.

### 4.1. Partitioning Logic
* Pick an arbitrary starting vertex and assign it to "Side 0".
* Assign all its neighbors to "Side 1".
* Assign the neighbors of those neighbors to "Side 0", and so on.
* Count how many vertices end up on each side.

## 5. Summary of Takeaways

* **Tree Characteristics**: Trees are always connected, acyclic, and bipartite. A tree with $n$ vertices always has exactly $n-1$ edges.
* **Bipartite Completion**: To maximize edges in a bipartite graph while keeping it simple, you must reach the state of a "Complete Bipartite Graph."
