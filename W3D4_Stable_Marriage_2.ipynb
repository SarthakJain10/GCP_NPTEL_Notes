# Implementation of the Stable Marriage Problem

This lecture focuses on the practical implementation details of the **Stable Marriage Problem** algorithm, commonly known as the Gale-Shapley algorithm. It covers data handling, logic for proposals and engagements, and helper functions necessary to manage the matching process.

---

## 1. Input Processing and Data Representation

The lecture highlights several nuances when dealing with input data from competitive programming platforms.

### 1.1 Managing IDs and Indices
* Men and women are typically represented by integers from **1 to N**.
* Since programming arrays (or vectors) are **zero-indexed** (0 to N-1), an offset of **±1** must be consistently applied when accessing or storing data.

### 1.2 Handling Preference Lists
* Each line of the preference input usually starts with a number indicating whose preference it is (e.g., the ID of the person).
* **Crucial Detail:** This leading number must be ignored or skipped so that it isn't mistaken for a ranking.
* The preferences for each person are stored in a collection of vectors (one for men, one for women).

### 1.3 Reversing Preferences
* The lecture suggests **reversing** the rankings in the data structure.
* **Logic:** By reversing, the top preference moves to the end of the array. This allows the algorithm to use "pop" operations to access and remove the current top choice efficiently as proposals are made.

## 2. The Core Algorithm Logic

The implementation centers around a main loop that continues as long as there is a man who is single and still has someone to propose to.

### 2.1 State Management
Two primary arrays are maintained to track the matching state :
* `Woman_Engagements`: Stores the current partner of each woman.
* `Man_Engagements`: Stores the current partner of each man.
* **Initialization:** All values are initially set to `-1` to signify that every person is currently single.

### 2.2 The Proposal Process
1. **Identify Single Men:** The algorithm iterates through the list of men to find one who is currently single.
2. **Select Top Choice:** The single man identifies the highest-ranked woman on his list to whom he hasn't proposed yet.
3. **Remove from List:** Once a proposal is made to a woman, she is removed from his list because he will never move "backward" or "up" his preference list later.

### 2.3 Acceptance and Displacement Logic
When a woman receives a proposal from a man (let's call him `Proposer`):
* **Case 1: The Woman is Single**
  * The match is made immediately.
* **Case 2: The Woman is Engaged**
  * The woman compares her `Current_Partner` with the `Proposer`.
  * If the `Proposer` is higher in her preference list (lower rank value), she breaks the old engagement.
  * The `Current_Partner` is returned to the "single" pool (set to `-1`), and she is now matched with the `Proposer`.
  * If she prefers her `Current_Partner`, no change is made, and the `Proposer` remains single for the next round.

## 3. Pseudocode for the Stable Marriage Implementation

The following pseudocode outlines the logical flow discussed in the lecture.

```python
FUNCTION SolveStableMarriage(N, MenPreferences, WomenPreferences):
    # Initialize all men and women as single
    Set ManEngagements array of size N to -1
    Set WomanEngagements array of size N to -1

    WHILE (SinglesRemaining(ManEngagements) is True):
        FOR each man I from 0 to N-1:
            IF ManEngagements[I] is -1 AND Man I still has preferences:
                # Identify the woman at the top of I's current list
                TopWoman = GetLastElement(MenPreferences[I])
                RemoveLastElement(MenPreferences[I])

                # Check if TopWoman is free
                IF WomanEngagements[TopWoman] is -1:
                    Match(I, TopWoman, ManEngagements, WomanEngagements)
                ELSE:
                    # She is engaged; compare current partner with new proposer
                    CurrentPartner = WomanEngagements[TopWoman]
                    
                    IF Rank(TopWoman, I) < Rank(TopWoman, CurrentPartner):
                        # New man is better; break old engagement
                        ManEngagements[CurrentPartner] = -1
                        Match(I, TopWoman, ManEngagements, WomanEngagements)
                    # ELSE: No change; man I remains single
    
    RETURN ManEngagements and WomanEngagements
```

## 4. Helper Functions

Several utility functions are used to keep the main logic clean:

* **`SinglesRemaining`**: Scans the engagement array to see if any value is still `-1`.
* **`GetRank`**: Retrieves the preference level of a specific man in a woman’s list. If the list was reversed, the rank calculation must account for the offset.
* **`Match`**: A simple utility to update both the `Man_Engagements` and `Woman_Engagements` arrays simultaneously to reflect a new pair.
