# Kruskal's Algorithm for Minimum Spanning Trees (MST)

This lecture provides an in-depth exploration of **Kruskal's Algorithm**, a global approach to finding the Minimum Spanning Tree (MST) of a connected, weighted graph. It contrasts this method with Prim's algorithm and details the implementation using the Disjoint Set Union (DSU) data structure.

---

## 1. Overview and Core Philosophy
Kruskal's algorithm builds an MST by taking a **global view** of the graph, rather than growing a single component from a starting vertex (as seen in Prim's algorithm).

* **Initial State**: The algorithm begins with a **spanning forest** where every vertex is an isolated component with no edges.
* **Iterative Process**: It iteratively picks **safe edges** from the entire graph based on their weights.
* **Goal**: To connect all vertices using the minimum total edge weight without forming cycles.


## 2. The Logic of "Safe" and "Useless" Edges
The algorithm processes edges in **increasing order of weight**.

### Safe Edges
An edge is considered **safe** if adding it to the current forest does not create a cycle. Because edges are processed from cheapest to most expensive, the first edge that connects two previously unconnected components is guaranteed to be a "safe" part of the MST.

### Useless Edges
An edge is **useless** if its endpoints already belong to the same component in the forest. Adding such an edge would create a **cycle**, violating the properties of a tree .

## 3. Algorithmic Steps (Simulation)
1.  List all edges in the graph.
2.  Sort these edges in ascending order based on their weights.
3.  Start with an empty set of edges.
4.  For each edge in the sorted list:
    * Check if the edge connects two different components (i.e., does it form a cycle?).
    * If no cycle is formed, add the edge to the MST and merge the two components.
    * If a cycle is formed, discard the edge.
5.  Stop when the MST contains **N - 1** edges (where N is the number of vertices).

## 4. Implementation Details
The algorithm is most efficiently implemented using a **Disjoint Set Union (DSU)** (or Union-Find) data structure.

### Data Structures Used
* **Edge List**: To store the graph as triplets containing the weight and the two connected vertices `(weight, u, v)` .
* **DSU (Union-Find)**: To keep track of connected components and efficiently check for cycles.
    * `find(u) == find(v)`: Checks if vertices `u` and `v` are in the same set (component).
    * `union(u, v)`: Merges the sets containing `u` and `v`.

### Pseudocode Representation
```python
ALGORITHM KruskalMST(Graph):
    mst_cost = 0
    num_edges_taken = 0
    mst_edges = empty list
    
    # Step 1: Initialize DSU where each vertex is its own set
    Initialize DSU with N elements
    
    # Step 2: Sort all edges by weight
    Sort(EdgeList)
    
    # Step 3: Process edges
    FOR each edge (weight, u, v) in EdgeList:
        # Check if u and v are in different components
        IF DSU.Find(u) is NOT equal to DSU.Find(v):
            # Edge is safe; add it to the MST
            mst_cost = mst_cost + weight
            mst_edges.Add(edge)
            DSU.Union(u, v)
            num_edges_taken = num_edges_taken + 1
            
            # Optimization: Stop early if we have N-1 edges
            IF num_edges_taken == (TotalVertices - 1):
                BREAK loop
                
    RETURN mst_cost, mst_edges
```

## 5. Performance and Complexity
The efficiency of Kruskal's algorithm is largely determined by the sorting step and the DSU operations.

* **Sorting**: Takes **O(M log M)** or **O(M log N)** time, where M is the number of edges.
* **DSU Operations**: With optimizations like *path compression* and *union by rank/depth*, these operations are near-constant (amortized).
* **Total Complexity**: Generally bounded by **O(M log N)** in the worst case.

## 6. Key Takeaways
* **Kruskal's vs. Prim's**: Kruskal's uses a global edge-sorting approach, while Prim's grows a tree from a single node. Both result in the same total MST cost.
* **Cycle Detection**: The core of the algorithm is identifying "useless" edges that create cycles using DSU.
* **Early Exit**: You can stop the algorithm as soon as you have collected **N - 1** edges, as any further edges will inevitably create cycles.
* **Connectivity**: If the graph is not connected, the algorithm will produce a **spanning forest** instead of a single tree
