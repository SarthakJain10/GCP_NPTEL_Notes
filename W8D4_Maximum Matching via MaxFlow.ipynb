# Maximum Matching via Maximum Flow

This lecture demonstrates how classic graph problems, specifically **Maximum Bipartite Matching**, can be modeled and solved using **Network Flow** techniques. The session uses a practical example to illustrate the conversion of a matching problem into a flow network.

---

## 1. Problem Overview: The "Bursu Ball"
The core objective is to find the maximum number of compatible, disjoint pairs between two distinct groups.

### The Scenario
* **Participants**: $N$ boys and $M$ girls.
* **Attribute**: Each person has a specific **dancing skill level** (an integer).
* **Compatibility Rule**: A boy and a girl can dance together only if their skill levels differ by **at most 1**.
* **Constraint**: Each person can be part of at most one pair (pairs must be **disjoint**).


---

## 2. Graph Representation
To solve this, the problem is first modeled as a graph:
* **Vertices**: Every boy and every girl is represented as a node.
* **Edges**: An edge exists between a boy and a girl if they are compatible.
* **Goal**: Find the **Maximum Matching**, which is the largest possible set of edges such that no two edges share a common vertex.

---

## 3. Modeling as a Flow Network
To use a Max-Flow algorithm (like Edmonds-Karp), the undirected matching graph must be transformed into a directed **Flow Network**.

### Construction Steps
1.  **Add a Source ($S$)**: Create an artificial starting node.
2.  **Add a Sink ($T$)**: Create an artificial terminal node.
3.  **Orient Edges**: Direct all edges from the "Girls" side to the "Boys" side.
4.  **Connect Source**: Add directed edges from $S$ to every girl.
5.  **Connect Sink**: Add directed edges from every boy to $T$.

### Setting Capacities
The choice of capacities is critical to enforce the "disjoint" constraint:
* **Source to Girls**: Capacity = **1**. This ensures each girl is used at most once.
* **Boys to Sink**: Capacity = **1**. This ensures each boy is used at most once.
* **Between Girls and Boys**: Capacity = **1** (though any value $\ge 1$ works as long as the source/sink constraints are unit-based).


---

## 4. Key Theoretical Insights
The lecture highlights two fundamental properties that make this reduction valid:

### Integral Flow Property
If all edge capacities in a network are integers, there exists a **Maximum Flow** where the flow through every edge is also an integer. In a unit-capacity network, this means every edge is either "fully used" (flow = 1) or "not used" (flow = 0).

### Equivalence of Flow and Matching
* **Flow $\rightarrow$ Matching**: A flow of value $K$ corresponds to a matching of size $K$. Because source/sink edges have capacity 1, the conservation of flow prevents any node from being matched more than once.
* **Matching $\rightarrow$ Flow**: Any valid matching of size $K$ can be converted into a valid flow of value $K$ by saturating the paths from $S \rightarrow \text{Girl} \rightarrow \text{Boy} \rightarrow T$ for each matched pair.

## 5. Algorithmic Approach (Pseudocode)

The following pseudocode outlines how to construct the network and compute the result.

```python
FUNCTION SolveMaximumMatching(boys_skills, girls_skills):
    # 1. Initialize Network
    num_boys = LENGTH(boys_skills)
    num_girls = LENGTH(girls_skills)
    source = 0
    sink = num_boys + num_girls + 1
    network = CREATE_FLOW_NETWORK(total_vertices = sink + 1)

    # 2. Add edges between compatible pairs
    FOR EACH girl_index FROM 1 TO num_girls:
        FOR EACH boy_index FROM 1 TO num_boys:
            IF ABS(girls_skills[girl_index] - boys_skills[boy_index]) <= 1:
                # Map girl and boy indices to unique vertex IDs
                ADD_DIRECTED_EDGE(network, FROM: girl_vertex, TO: boy_vertex, CAPACITY: 1)

    # 3. Connect Source and Sink
    FOR EACH girl_vertex:
        ADD_DIRECTED_EDGE(network, FROM: source, TO: girl_vertex, CAPACITY: 1)
    
    FOR EACH boy_vertex:
        ADD_DIRECTED_EDGE(network, FROM: boy_vertex, TO: sink, CAPACITY: 1)

    # 4. Compute Max Flow
    max_matching_size = COMPUTE_MAX_FLOW(network, source, sink)
    
    RETURN max_matching_size
```

## 6. Summary of Takeaways
* **Network Flow Versatility**: Max-Flow is not just for "pipes and water"; it is a powerful modeling tool for constraints.
* **Bipartite Matching**: Finding the largest matching in a bipartite graph is equivalent to finding the Max-Flow in a unit-capacity network structured with a source and sink.
* **Implementation Strategy**: When solving such problems, the "heavy lifting" is done by standard algorithms (like Edmonds-Karp); the programmer's primary task is the correct **construction and modeling** of the network.
* **Efficiency Note**: While Max-Flow works, specialized bipartite matching algorithms exist for extremely large constraints where general flow algorithms might be slower.
