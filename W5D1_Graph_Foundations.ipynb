# Graph Foundations: Representation and Traversal

This lecture introduces the fundamental concepts of graph theory as applied in competitive programming, focusing on how graphs are defined, stored in memory, and the conceptual intuition behind exploring them.

---

## 1. Defining a Graph
A graph is formally represented as a pair **(V, E)**.
* **V (Vertices/Nodes):** A set of elements representing entities (e.g., locations, people, webpages).
* **E (Edges/Connections):** A collection of pairwise relationships between these vertices.

### Types of Graphs
* **Undirected Graph:** Relationships are mutual or symmetric (e.g., Facebook friendships). Edges have no specific direction.
* **Directed Graph:** Relationships are asymmetric (e.g., Twitter followers). Edges are represented by arrows showing the direction of the relationship.
* **Weighted Graph:** Edges carry additional information, such as cost, distance, or time (e.g., road lengths in a navigation system).
* **Simple Graph:** A graph that does not contain multiple edges between the same two vertices or "self-loops" (an edge from a vertex to itself).
* **Multi-graph:** Allows for multiple edges between the same pair of vertices or self-loops.

## 2. Graph Representations in Code
Choosing the right data structure is critical for staying within time and memory limits in competitive programming.

### A. Adjacency Matrix
An **N x N matrix** where N is the number of vertices. 
* An entry at row `i` and column `j` is set to **1** (or the **weight**) if an edge exists from `i` to `j`, and **0** otherwise.
* **Pros:** Very fast for checking if a specific edge exists between two nodes (O(1) lookup).
* **Cons:** Space-intensive (**N²** memory). If you have 100,000 vertices, an adjacency matrix is infeasible. It is also slow (O(N)) for finding all neighbors of a single node because you must scan the entire row.

### B. Adjacency List
A collection of lists where each vertex has its own list containing its neighbors.
* For weighted graphs, the list stores pairs: `(neighbor_label, weight)`.
* **Pros:** Space-efficient; memory usage is proportional to the number of edges. It is very fast for iterating through the neighbors of a specific node (proportional to the node's **degree**).
* **Cons:** Slower to check if a specific edge exists between two nodes (must scan the list).

## 3. Graph Traversal Intuition
Traversal is the process of visiting all nodes in a graph. The lecture explains the logic through "web browsing" analogies.

### Depth First Search (DFS): "The Rabbit Hole"
DFS follows a path as deep as possible before backtracking.
* **Logic:** Start at a node, pick an unvisited neighbor, and immediately move to it. Repeat this recursively.
* **Backtracking:** When you hit a "dead end" (a node with no unvisited neighbors), go back to the previous node and look for other unvisited paths.

**Pseudocode for DFS Logic:**
```
Function ExploreDepthFirst(CurrentNode):
    Mark CurrentNode as visited
    For each Neighbor of CurrentNode:
        If Neighbor is not visited:
            ExploreDepthFirst(Neighbor)
```

### Breadth First Search (BFS): "The Layered Approach"
BFS explores the graph level by level, visiting all immediate neighbors before moving to the neighbors' neighbors.
* **Logic:** Use a **Queue** data structure. Visit the starting node, then put all its unvisited neighbors into a queue.
* **Processing:** Take the first node out of the queue, visit its unvisited neighbors, and add them to the end of the queue.

**Pseudocode for BFS Logic:**
```
Function ExploreBreadthFirst(StartNode):
    Create an empty Queue
    Add StartNode to Queue and mark as visited
    
    While Queue is not empty:
        CurrentNode = Remove first item from Queue
        For each Neighbor of CurrentNode:
            If Neighbor is not visited:
                Mark Neighbor as visited
                Add Neighbor to the end of Queue
```

---

## Summary of Key Takeaways
* **Graph Choice:** Most competitive programming problems use **Simple, Undirected, and Unweighted** graphs as a baseline, but the concepts extend to directed and weighted versions.
* **Storage:** Use **Adjacency Matrices** for small graphs (up to ~3,000 vertices) when edge-lookup speed is vital. Use **Adjacency Lists** for large, sparse graphs to save space and traverse neighbors efficiently.
* **Traversal:** **DFS** is naturally recursive and focuses on depth, while **BFS** uses a queue to explore the graph in layers (useful for finding the shortest path in unweighted graphs).
