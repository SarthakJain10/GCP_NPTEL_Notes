# Sam I AM: Solving Vertex Covers using Network Flows

This lecture explores the "Sam I AM" problem from the UVA platform, demonstrating how to solve a grid-based optimization problem by modeling it as a **Minimum Vertex Cover** in a **Bipartite Graph** and solving it using the **Max-Flow Min-Cut Theorem**.

---

## 1. Problem Overview: Sam I AM
The problem involves a rectangular grid (temple) containing enemies at specific coordinates $(row, column)$. The goal is to eliminate all enemies using the minimum number of cannonballs.

### Core Rules:
* A single cannonball shot can destroy all enemies in an entire **row** or an entire **column**.
* **Objective:** Find the minimum total number of row and column shots required to cover every cell containing an enemy.


---

## 2. Modeling the Problem as a Graph
To solve this, the spatial grid is transformed into a graph structure where the interaction between rows and columns becomes the focus.

### Graph Construction:
* **Vertices:** Create a set of vertices for every row ($R_1, R_2, \dots$) and every column ($C_1, C_2, \dots$).
* **Edges:** Add an edge between row vertex $R_i$ and column vertex $C_j$ if and only if there is an enemy at the intersection of row $i$ and column $j$.
* **Bipartite Nature:** This graph is **bipartite** because edges only exist between a row vertex and a column vertex; no two row vertices or two column vertices are connected to each other.

### The Vertex Cover Concept:
In graph theory, a **Vertex Cover** is a subset of vertices such that every edge in the graph is incident to at least one vertex in the subset.
* In this problem:
    * **Edges** = Enemy locations.
    * **Vertices** = Possible cannonball shots (rows or columns).
* Therefore, the minimum number of cannonballs is equivalent to the **Minimum Vertex Cover** of this bipartite graph.

---

## 3. The Duality: Matching vs. Vertex Cover
A key insight provided is the relationship between "Independent Sets of Enemies" and "Matchings".

* **Matching:** A set of edges that share no common vertices.
* **Lower Bound:** If you have three enemies that share no common row or column (a matching of size 3), you **must** use at least three cannonballs because no single shot can hit more than one of them.
* **Konig’s Theorem:** In any bipartite graph, the number of edges in a **Maximum Matching** is exactly equal to the number of vertices in a **Minimum Vertex Cover**.

---

## 4. Solving via Max-Flow Min-Cut
To find the specific rows and columns for the optimal solution, the bipartite graph is converted into a flow network.

### Flow Network Setup:
1. **Source ($S$) and Sink ($T$):** Add a source $S$ and a sink $T$.
2. **Source Edges:** Add directed edges from $S$ to every row vertex ($R_i$) with **capacity 1**.
3. **Sink Edges:** Add directed edges from every column vertex ($C_j$) to $T$ with **capacity 1**.
4. **Internal Edges:** Orient original edges from row vertices to column vertices ($R_i \to C_j$) with **infinite capacity**.


### Finding the Minimum Cut:
After running a Max-Flow algorithm (like Edmonds-Karp or Dinic), we find the **Minimum Cut** by identifying the set of vertices reachable from $S$ in the **residual graph**.

---

## 5. Identifying the "Misfits" (The Solution)
The lecture introduces the concept of **"Misfits"** to identify which specific rows or columns belong to the Minimum Vertex Cover.

* **Row Misfits:** Row vertices that are **NOT reachable** from $S$ in the residual graph.
* **Column Misfits:** Column vertices that **ARE reachable** from $S$ in the residual graph.

These "Misfit" vertices constitute the minimum vertex cover. The total number of these vertices will equal the Max-Flow value.

---

## 6. Pseudocode Implementation
The following logic describes how to extract the optimal shots after computing the flow.

```markdown
ALGORITHM FindMinimumShots:
    1. Construct Bipartite Graph from enemy locations (Row nodes to Column nodes)
    2. Build Flow Network:
        - S -> Row nodes (capacity 1)
        - Column nodes -> T (capacity 1)
        - Row nodes -> Column nodes (capacity infinity)
    3. Run Max-Flow algorithm from S to T
    4. Find ReachableSet:
        - Use BFS/DFS on the residual graph starting from S
        - Mark all nodes reached as 'Reachable'
    
    5. Identify Optimal Shots:
        FOR each row_node in AllRows:
            IF row_node is NOT in ReachableSet:
                Output "Shoot Row " + row_node_index
        
        FOR each col_node in AllColumns:
            IF col_node IS in ReachableSet:
                Output "Shoot Column " + col_node_index
```

## Summary of Takeaways
* **Problem Transformation:** Problems involving "covering" a grid with lines can often be modeled as **Minimum Vertex Cover** in bipartite graphs.
* **Konig’s Theorem:** For bipartite graphs, **Max Matching = Min Vertex Cover**.
* **Flow Utility:** The **Max-Flow** value gives the minimum number of shots, and the **Min-Cut** (reachable vs. non-reachable nodes in the residual graph) identifies exactly which rows and columns to shoot.
* **Complexity:** This approach is more efficient than exhaustive searching and provides a guaranteed optimal solution for competitive programming
