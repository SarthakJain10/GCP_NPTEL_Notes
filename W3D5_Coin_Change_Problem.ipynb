# When Greedy Does Not Work: The Coin Change Problem

This lecture explores the **Greedy Strategy** in the context of the **Coin Change Problem**, specifically examining why it works for certain currency systems and fails for others. It serves as an introduction to understanding the limitations of greedy algorithms and the need for more sophisticated approaches in specific scenarios.

---

## 1. Introduction to the Coin Change Problem

The Coin Change Problem is a classic computational puzzle involving the generation of a specific total amount using a set of available currency denominations.

### Problem Definition:
* **Objective**: Generate a total amount `N` using the **minimum** number of notes/coins.
* **Constraints**:
    * You have an unlimited supply of each denomination.
    * Denominations are predefined (e.g., $1, $5, $10, $20, $100).
* **Goal**: Minimize the total count of physical bills or coins used to reach the target sum.

---

## 2. The Greedy Strategy for Coin Change

The "natural" or intuitive way to solve this problem is through a **Greedy Approach**. In a greedy algorithm, you make the best possible choice at each immediate step without worrying about future consequences.

### The Greedy Logic:
At every step, choose the **largest possible denomination** that is less than or equal to the remaining amount to be generated.

### Pseudocode Representation:
```python
FUNCTION GreedyCoinChange(TotalAmount, Denominations):
    Sort Denominations in descending order
    Count = 0
    Remaining = TotalAmount
    
    WHILE Remaining is greater than 0:
        Find the largest denomination D such that D <= Remaining
        IF no such D exists:
            The amount cannot be formed (Return Error)
        
        Use as many of denomination D as possible:
        NumberToUse = Floor(Remaining / D)
        Count = Count + NumberToUse
        Remaining = Remaining - (NumberToUse * D)
        
    RETURN Count
END FUNCTION
```

### Example 1 (Success):
* **Target**: $125
* **Denominations**: $100, $50, $20, $10, $5, $1
* **Steps**:
    1.  Pick the largest: $100 (Remaining: $25)
    2.  Next largest possible: $20 (Remaining: $5)
    3.  Next largest possible: $5 (Remaining: $0)
* **Total notes**: 3 (Optimal)

## 3. Canonical Coin Systems and Divisibility

The reason the greedy approach works for standard currencies (like the US Dollar or Indian Rupee) is that they are **Canonical Coin Systems**.

### The Divisibility Property:
In these systems, a larger denomination is usually a multiple of the smaller denominations below it (or follows a specific mathematical property).
* **Example**: $100 is divisible by $20, which is divisible by $10, which is divisible by $5.
* **Intuition**: Because of this divisibility, if you can form an amount using smaller bills, you can always "upgrade" them to a single larger bill to reduce the total count. 


---

## 4. When Greedy Fails (Non-Canonical Systems)

The greedy approach is not globally optimal for all sets of denominations. If the denominations do not share specific mathematical relationships (like the divisibility property), the greedy choice at the start might prevent you from reaching the most efficient solution later.

### Example 2 (Failure):
* **Target**: 6 units
* **Denominations**: 4, 3, 1
* **Greedy Approach**:
    1.  Pick the largest: **4** (Remaining: 2)
    2.  Next largest possible: **1** (Remaining: 1)
    3.  Next largest possible: **1** (Remaining: 0)
    * **Total units**: 3 (4 + 1 + 1)
* **Optimal Approach (Non-Greedy)**:
    1.  Pick **3** (Remaining: 3)
    2.  Pick **3** (Remaining: 0)
    * **Total units**: 2 (3 + 3)

**Observation**: The greedy algorithm was "too greedy" by picking the 4 first, which forced it to use two more 1s. By picking a smaller number (3) first, we achieved a better overall result.

---

## 5. Conceptual Summary of Findings

* **Greedy Choice**: A greedy algorithm makes the locally optimal choice at each stage (picking the biggest bill).
* **Optimality**: In standard coin systems, the locally optimal choice leads to a globally optimal solution.
* **Limitations**: In "arbitrary" or non-canonical systems, the greedy choice can be a "trap" that leads to a sub-optimal total count.
* **Conclusion**: To solve the coin change problem for *any* set of denominations, more advanced techniques (like Dynamic Programming) are required, which the lecture hints at for future modules.


---
**Main Takeaway**: While the greedy strategy is intuitive and fast, its correctness depends entirely on the structure of the data (the denominations). If the "divisibility" or canonical property is broken, greedy logic fails to provide the minimum number of coins.
