# Graph Dominating Sets: The "Cover It!" Problem

This lecture explores an application of **Breadth-First Search (BFS)** to solve a graph theory problem involving "dominating sets." The primary objective is to find a subset of vertices that "covers" the entire graph under specific constraints.


## 1. Problem Definition and Constraints

The problem, titled **"Cover It!"**, involves a specific type of graph and a requirement for vertex selection.

* **Graph Type:** An undirected, unweighted, and **connected** simple graph (no self-loops or multiple edges) with $n$ vertices and $m$ edges.
* **The Task:** Find a subset of vertices $S$ such that:
    1.  The size of $S$ is at most $\lfloor n/2 \rfloor$ (at most half the total vertices).
    2.  Every vertex not in $S$ is adjacent to at least one vertex that *is* in $S$.
* **Terminology:** In graph theory, such a subset $S$ is known as a **Dominating Set**.
* **Guarantee:** A solution is guaranteed to exist for any connected graph of this type.

## 2. The Core Observation: Layering with BFS

The key to solving this problem efficiently is observing how a **Breadth-First Search (BFS)** organizes a graph into layers based on distance from a starting source vertex.

### Odd and Even Layers
When performing a BFS starting from an arbitrary vertex (root), the vertices are discovered in levels:
* **Layer 0 (Even):** The starting vertex.
* **Layer 1 (Odd):** Direct neighbors of the starting vertex.
* **Layer 2 (Even):** Neighbors of Layer 1 vertices (not already visited).
* **Layer 3 (Odd):** Neighbors of Layer 2 vertices (not already visited), and so on.

### Why Layers Form Dominating Sets
1.  **Red (Even) Partition:** Every vertex in an odd layer was discovered by a "parent" in an even layer. Thus, every "Odd" vertex has at least one "Even" neighbor. 
2.  **Blue (Odd) Partition:** Every vertex in an even layer (except the root) was discovered by a "parent" in an odd layer. The root vertex itself is connected to its children in the first odd layer.
3.  **Conclusion:** Both the set of all vertices in **even layers** and the set of all vertices in **odd layers** are valid dominating sets.


## 3. Algorithm Logic

Since the goal is to find a dominating set of size $\leq \lfloor n/2 \rfloor$, we simply partition the graph into these two sets (Even and Odd) and pick the smaller one. Because the two sets together contain all $n$ vertices, the smaller of the two must have a size less than or equal to $n/2$.

### Pseudocode Implementation

```pseudocode
FUNCTION FindSmallDominatingSet(Graph, StartVertex):
    // Initialize structures
    Initialize Queue and add StartVertex
    Initialize DistanceArray where all distances = -1
    DistanceArray[StartVertex] = 0
    
    // Lists to store vertices by layer parity
    Initialize List EVEN_VERTICES
    Initialize List ODD_VERTICES
    Add StartVertex to EVEN_VERTICES

    WHILE Queue is not empty:
        CurrentNode = Queue.RemoveFront()
        
        FOR EACH Neighbor of CurrentNode:
            IF DistanceArray[Neighbor] is -1 (Not Visited):
                // Calculate distance and add to queue
                DistanceArray[Neighbor] = DistanceArray[CurrentNode] + 1
                Queue.AddBack(Neighbor)
                
                // Assign to appropriate set based on distance
                IF DistanceArray[Neighbor] is EVEN:
                    Add Neighbor to EVEN_VERTICES
                ELSE:
                    Add Neighbor to ODD_VERTICES
    
    // Select and return the smaller set
    IF Size(EVEN_VERTICES) <= Size(ODD_VERTICES):
        RETURN EVEN_VERTICES
    ELSE:
        RETURN ODD_VERTICES
END FUNCTION
```

## 4. Summary of Main Takeaways

* **Bipartite-like Partitioning:** Even if a graph is not bipartite (e.g., it contains triangles), BFS distance-based coloring creates a partition that behaves similarly for domination purposes.
* **Parent-Child Relationship:** The fundamental property of BFS—that every node at distance $d$ has a neighbor at distance $d-1$—guarantees that the parity-based sets (Odd/Even) cover each other.
* **Efficiency:** This approach is highly efficient as it only requires a single BFS traversal, resulting in a time complexity of $O(V + E)$, where $V$ is vertices and $E$ is edges.
* **Universality:** The lecture proves that any connected, undirected graph always contains a dominating set that is at most half its size.
