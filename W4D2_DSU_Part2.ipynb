# Disjoint Set Union (DSU) via Union by Rank and Path Compression

This lecture focuses on the implementation and optimization of the **Disjoint Set Union (DSU)** data structure. The goal is to efficiently manage a collection of disjoint sets and support two primary operations: finding the representative of a set and merging two sets.

---

## 1. Fundamental Concepts and Operations

The DSU data structure operates over a fixed universe of **n** elements. Each set within this universe is identified by a unique **representative element** (also called a leader).

### Primary Operations:
* **`FindSet(i)`**: Takes an element `i` and returns the representative of the set to which `i` belongs.
* **`UnionSet(i, j)`**: Merges the two sets containing elements `i` and `j`. If they already belong to the same set, no action is taken.

---

## 2. Naive Implementation: Representative Array

The simplest way to implement DSU is using a **Leader Array**, where the index represents the element and the value at that index stores its representative.

### Characteristics:
* **Initialization**: Every element starts in its own singleton set, meaning `Leader[i] = i`.
* **FindSet**: Performed in **O(1)** (constant time) by simply looking up the array index.
* **UnionSet**: To merge sets with representatives `x` and `y`, the algorithm must scan the *entire* array and change all occurrences of `x` to `y` (or vice-versa). This results in **O(n)** (linear time) complexity.

---

## 3. Tree-Based Implementation (Pointer Approach)

To improve efficiency, elements can be viewed as **pointers**. Each element points to its "parent" in a tree structure. The root of the tree is the **representative** of the set and points to itself.

### The Mechanics:
* **`FindSet(i)`**: Follow the parent pointers from element `i` upwards until reaching an element that points to itself (the root).
* **`UnionSet(i, j)`**: Find the roots of the trees containing `i` and `j`. If the roots are different, point the root of one tree to the root of the other.


### Complexity Issue:
In a naive tree implementation, trees can become very tall (skewed), leading to a **FindSet** complexity of **O(n)** in the worst case.

---

## 4. Optimization: Union by Rank (Depth)

To prevent trees from becoming too deep, the **Union by Rank** heuristic ensures that the "shallower" tree is always attached to the root of the "deeper" tree.

### Logic:
* Keep track of the **rank** (an upper bound on the height/depth) of each tree.
* When merging two trees with depths `d1` and `d2`:
    * If `d1 < d2`: Attach tree 1 to the root of tree 2. The overall depth remains `d2`.
    * If `d1 > d2`: Attach tree 2 to the root of tree 1. The overall depth remains `d1`.
    * If `d1 == d2`: Attach one to the other and increment the rank of the new root by 1.

**Result**: This ensures that the height of any tree is bounded by **O(log n)**.

---

## 5. Optimization: Path Compression

**Path Compression** is a heuristic applied during the `FindSet` operation to flatten the structure of the tree.

### Process:
While traversing the path from an element to the root, every node encountered is updated to point **directly to the root**. This significantly speeds up future `FindSet` operations for those nodes and their descendants.

### Recursive Logic (Pseudocode):
```
Function FindSet(i):
    // Base Case: if i is its own parent, it is the root
    If Parent[i] == i:
        Return i
    Else:
        // Recursive Step: Path Compression
        // Update Parent[i] to point directly to the root found via recursion
        Parent[i] = FindSet(Parent[i])
        Return Parent[i]
```

---

## 6. Combined Complexity

When **Union by Rank** and **Path Compression** are used together, the efficiency of the operations improves dramatically.

* **Amortized Time**: While a single operation might occasionally take longer, a sequence of `m` operations on `n` elements takes nearly linear time.
* **Inverse Ackermann Function**: The complexity is formally expressed as **O(α(n))** per operation, where **α** is the inverse Ackermann function. This function grows so slowly that for all practical purposes (even with billions of elements), it is considered **constant time O(1)**.

---

## Summary of Key Takeaways

* **Naive DSU** using a flat array results in fast "Find" but slow "Union" (O(n)).
* **Tree-based DSU** allows for more flexibility but can lead to inefficient, tall trees.
* **Union by Rank** limits tree height to logarithmic scale by always attaching smaller trees to larger ones.
* **Path Compression** flattens trees during "Find" operations, making the structure increasingly efficient over time.
* **Combined**, these optimizations provide **near-constant amortized time complexity**, making DSU highly effective for tasks like maintaining connected components in a graph.
