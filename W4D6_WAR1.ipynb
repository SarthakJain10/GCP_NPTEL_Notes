# Analysis of the "War" Problem (Disjoint Set Union Application)

This lecture explores a complex application of the **Disjoint Set Union (DSU)** data structure through a problem titled "War." The problem involves managing relationships between people based on friendship and rivalry rules.


## 1. Problem Overview
The core of the problem involves $N$ people, each belonging to one of two different countries. 
* **Same Country:** People from the same country are **friends**.
* **Different Countries:** People from different countries are **enemies** (or rivals).

### The Challenge
Initially, you do not know which person belongs to which country. You must infer these relationships based on a series of observations (operations) and answer questions (queries) about the status of two people.

## 2. The Three Rules of Inference
Beyond direct observations, relationships can be deduced using three fundamental rules:

### Rule 1: Transitivity of Friendship
Friendship is transitive. If person **A** is friends with **B**, and **B** is friends with **C**, then **A** is also friends with **C**.
* **Logic:** "The friend of my friend is my friend."

### Rule 2: Common Enemy Rule
If two people share the same enemy, they must belong to the same country and are therefore friends.
* **Logic:** "A common enemy makes two people friends."

### Rule 3: Enemy of a Friend
If your friend has an enemy, that person is also your enemy because they must be from the opposing country.
* **Logic:** "An enemy of a friend is an enemy."


## 3. Operations and Queries
The program must handle a stream of input consisting of four types of actions involving two people ($P$ and $Q$):

| Action | Description | Expected Output |
| :--- | :--- | :--- |
| **setFriends(P, Q)** | Record that $P$ and $Q$ are friends. | None (or `-1` if it contradicts facts) |
| **setEnemies(P, Q)** | Record that $P$ and $Q$ are enemies. | None (or `-1` if it contradicts facts) |
| **areFriends(P, Q)** | Check if $P$ and $Q$ are friends. | `1` (Yes) or `0` (No) |
| **areEnemies(P, Q)** | Check if $P$ and $Q$ are enemies. | `1` (Yes) or `0` (No) |

### Handling Contradictions
If a `setFriends` or `setEnemies` operation contradicts information already established (either directly or by inference), the operation must be ignored, and the program should output `-1`.

## 4. Conceptual Logic & Pseudocode
The problem can be modeled using **Disjoint Sets** because friendships form clusters. When two people from different clusters become friends, their entire clusters merge.

### Friendship Logic (Union-Find)
To check if two people are friends, we determine if they belong to the same "friendship cluster" using the `find` operation.

```python
Function AreFriends(Person_P, Person_Q):
    If FindRoot(Person_P) == FindRoot(Person_Q):
        Return True
    Else:
        Return False
```
