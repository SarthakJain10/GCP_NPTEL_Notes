# Islands War: Greedy Algorithm Analysis

This lecture explores a specific problem-solving scenario known as the **Islands War** (from AtCoder contest 103). It focuses on applying a **Greedy Algorithm** to resolve conflicts between islands by removing the minimum number of bridges.

## 1. Problem Definition
The scenario consists of $N$ islands arranged linearly from West to East, numbered 1 to $N$.

* **Connectivity**: There are $N-1$ bridges. Bridge $i$ connects Island $i$ with Island $i+1$.
* **The Conflict**: There are $M$ requests (disputes). Each request is a pair $(a_i, b_i)$, indicating that Island $a_i$ and Island $b_i$ must be made unable to reach each other.
* **The Task**: Minimize the total number of bridges removed to satisfy all $M$ requests. Removing a bridge between two islands breaks the path between them and any other islands that relied on that specific bridge for travel.

## 2. Failed Greedy Strategy: The "Most Powerful Bridge"
A natural but incorrect first instinct is to target the bridge that satisfies the most requests simultaneously.

### Strategy Description
1.  **Define "Power"**: The power of a bridge is the number of active requests that would be satisfied if that bridge were removed.
2.  **Selection**: Sort bridges by power and remove the one with the highest power.
3.  **Update**: Remove all requests satisfied by that bridge and recompute the power for remaining bridges.
4.  **Repeat**: Continue until all requests are satisfied.

### Why It Fails
This strategy is "locally" optimal but "globally" inefficient. In certain configurations, picking the most powerful bridge early can force the removal of more bridges later than if a less "powerful" bridge had been chosen first .
* **Counter-example**: A configuration of many disjoint requests and some overlapping requests can lead the algorithm to pick a bridge that handles the overlaps but fails to efficiently handle the disjoint segments, resulting in a sub-optimal total count.


## 3. The Optimal Greedy Strategy: Right-Endpoint Sorting
The correct approach focuses on the **constraints** imposed by the requests, specifically looking at where requests "end" (the easternmost point).

### The Main Insight
To satisfy a request $(a_i, b_i)$, at least one bridge between $a_i$ and $b_i$ must be removed. To maximize the utility of a single removal, we should pick a bridge that is as far to the "right" (East) as possible within that request's range, because a right-positioned bridge is more likely to overlap with future requests that start later.

### The Algorithm Steps
1.  **Sort**: Sort all requests $(a_i, b_i)$ based on their **right endpoint** ($b_i$) in ascending order.
2.  **Initialize**: Set a variable to track the position of the `last_removed_bridge`.
3.  **Iterate**: Process each sorted request:
    * If the current request $(a_i, b_i)$ is **not already satisfied** by the `last_removed_bridge` (i.e., the bridge was removed to the left of $a_i$):
        * Remove the bridge immediately to the left of $b_i$ (the bridge connecting $b_i - 1$ and $b_i$).
        * Increment the total count.
        * Update `last_removed_bridge` to this new position.
    * If the request is already satisfied, skip it.

### Pseudocode Representation
```markdown
FUNCTION minimize_bridge_removal(islands_count, requests_list):
    SORT requests_list by their right_endpoint (ascending)
    
    SET total_removed_bridges = 0
    SET last_bridge_position = -1
    
    FOR EACH request(start, end) IN requests_list:
        // Check if the current dispute is already disconnected
        IF last_bridge_position < start:
            // Remove the rightmost possible bridge for this request
            // This bridge is between (end - 1) and (end)
            last_bridge_position = end - 1
            total_removed_bridges = total_removed_bridges + 1
            
    RETURN total_removed_bridges
```

## 4. Correctness and Complexity
### Proof of Correctness (Duality)
The algorithm effectively identifies the **maximum number of disjoint requests**. Since each disjoint request requires at least one unique bridge removal, and our algorithm satisfies all requests using exactly that many bridges, the solution must be optimal.

### Complexity Analysis
* **Sorting**: $O(M \log M)$, where $M$ is the number of requests.
* **Linear Pass**: $O(M)$ to iterate through the requests once.
* **Overall**: $O(M \log M)$, which is well within the limits for $M = 10^5$ .

## 5. Summary of Takeaways
* **Greedy Choice**: The "most powerful" item isn't always the best greedy choice; sometimes the "most constrained" item (the one that ends first) provides the correct priority.
* **Sorting Matters**: In interval-based problems, sorting by end-times/right-endpoints is a common and powerful technique.
* **Efficiency**: By tracking only the `last_bridge_position`, we avoid $O(M^2)$ complexity and achieve a highly efficient $O(M \log M)$ solution .
