# Diamond Inheritance and Directed Graph Traversals

This lecture explores the application of graph traversal algorithms—specifically **Depth First Search (DFS)**—to solve the "Diamond Inheritance" problem, a concept derived from object-oriented programming.

## 1\. The Concept of Diamond Inheritance

In object-oriented programming, a **Diamond Inheritance** (sometimes called the "Deadly Diamond of Death") occurs when there is ambiguity in the inheritance hierarchy.

### 1.1 Formal Definition

A class diagram has diamond inheritance if there exists a pair of classes $X$ and $Y$ such that there are at least **two different inheritance paths** from $X$ to $Y$.

### 1.2 Mapping to Graph Theory

To solve this computationally, the problem is modeled as a **Directed Graph**:

  * **Vertices (Nodes):** Represent classes.
  * **Directed Edges:** An edge from vertex $A$ to vertex $B$ indicates that class $A$ inherits from class $B$.
  * **Inheritance Path:** A sequence of directed edges connecting two vertices.

## 2\. Analyzing Path Differences

The lecture clarifies what it means for two paths between $X$ and $Y$ to be "different":

1.  **Internally Disjoint Paths:** The paths share no vertices except for the start $(X)$ and the end $(Y)$.
2.  **Distinct Paths:** The paths differ by at least one vertex.

**Equivalence Logic:** The lecture argues that if two paths differ by at least one vertex, you can always find a sub-segment of those paths that is internally disjoint. Therefore, detecting any two distinct paths between the same two nodes is sufficient to confirm a diamond inheritance.

## 3\. Detection Strategy using DFS

The core idea is to identify if any vertex can be reached more than once starting from the same source.

### 3.1 Key Claims

  * **Claim 1:** If a DFS traversal starting from a source vertex $X$ visits a vertex $Y$ that has already been marked as "visited," then a diamond inheritance exists.
  * **Claim 2:** If a diamond inheritance exists between $X$ and $Y$, then a DFS starting from $X$ will inevitably attempt to visit $Y$ more than once.

### 3.2 Algorithm Logic

Because the diamond can start at any node, the algorithm must attempt a traversal starting from **every** vertex in the graph.

#### Pseudocode for Diamond Detection

```python
FUNCTION detect_diamond(graph, num_vertices):
    FOR each vertex i from 1 to num_vertices:
        # Reset state for each new potential source
        is_diamond_found = FALSE
        visited_array = array of size num_vertices, initialized to FALSE
        
        # Start DFS from vertex i
        run_dfs(i, visited_array, is_diamond_found)
        
        IF is_diamond_found is TRUE:
            RETURN "Yes, Diamond Inheritance Found"
            
    RETURN "No Diamond Inheritance Found"

FUNCTION run_dfs(current_vertex, visited_array, is_diamond_found):
    # If we already found a diamond in this branch, stop
    IF is_diamond_found is TRUE:
        RETURN

    # Mark current vertex as visited
    visited_array[current_vertex] = TRUE

    FOR each neighbor in graph.get_neighbors(current_vertex):
        IF visited_array[neighbor] is TRUE:
            # We reached a vertex that was already visited in this traversal!
            is_diamond_found = TRUE
            RETURN
        ELSE:
            run_dfs(neighbor, visited_array, is_diamond_found)
```

## 4\. Implementation Considerations

  * **Source Exhaustion:** You cannot simply start DFS from a single arbitrary vertex. In a directed graph, if you start at a "sink" node (a node with no outgoing edges), the DFS will terminate immediately, potentially missing diamonds elsewhere in the graph.
  * **State Reset:** It is crucial to clear the `visited` array and reset flags before starting a search from a new source vertex.
  * **Efficiency:** While the approach involves multiple DFS passes, it effectively covers all possible inheritance paths to identify duplicates.

## 5\. Summary of Main Takeaways

  * **Diamond inheritance** is defined as the existence of multiple directed paths between any two nodes $X$ and $Y$.
  * **Directed graphs** are the ideal model for representing inheritance hierarchies.
  * **DFS (or BFS)** can be adapted to detect diamonds by checking if a vertex is encountered more than once during a single traversal session.
  * **Global Search:** To ensure correctness in directed graphs, the traversal must be initiated from every vertex in the graph to account for different starting points of potential diamonds.
