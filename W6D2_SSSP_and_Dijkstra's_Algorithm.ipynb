# Single Source Shortest Path (SSSP) and Dijkstra's Algorithm

This lecture focuses on solving the **Single Source Shortest Path (SSSP)** problem in graphs with **positive edge weights**. It builds upon the concepts of Breadth-First Search (BFS) and explores how to efficiently find the shortest path without the expensive overhead of modifying the graph structure.

---

## 1. Context and Problem Definition

### Limitations of Breadth-First Search (BFS)
* **Unweighted Graphs:** BFS works perfectly for finding the shortest path when all edge weights are equal to 1.
* **Weighted Graphs:** BFS fails to identify the shortest path once different edge weights are introduced.
* **The "Puncturing" Workaround:** While one can simulate weights by adding artificial intermediate vertices to "stretch" edges, this is computationally expensive and inefficient.

### Objective
To find an algorithm that predicts the behavior of BFS on a weighted graph (specifically with positive weights) but operates directly on the original graph for better efficiency.

---

## 2. Key Concepts: Tense Edges and Relaxation

The algorithm relies on identifying and resolving "tension" in the graph's edges.

### The Tense Edge
An edge from vertex **u** to vertex **v** with weight **w** is considered **tense** if the current known distance to **v** can be improved by going through **u** .

* **Mathematical Condition:** An edge is tense if:
  `Distance(u) + weight(u, v) < Distance(v)`
* **Interpretation:** This inequality indicates that the current stored distance for vertex **v** is outdated or sub-optimal.
* **Edge Cases:** * If `Distance(u)` is infinity, the edge is not tense because we don't yet know how to reach **u**.
    * At the start, all edges originating from the source vertex are tense because the source distance is 0 and all others are infinity.


### Relaxation
**Relaxation** is the process of "resolving" the tension of an edge by updating the distance to the destination vertex.
* **Action:** Update `Distance(v)` to the new, smaller value: `Distance(u) + weight(u, v)`.
* **Predecessor Tracking:** To reconstruct the actual path, the algorithm updates a "predecessor pointer" for vertex **v**, setting it to **u**.

---

## 3. Dijkstra's Algorithm Logic

Dijkstra's algorithm systematically extracts the closest vertex to the source and relaxes its outgoing edges.

### Initialization
1. Set the distance of the **Source Vertex** to **0**.
2. Set the distance of all **other vertices** to **Infinity** (or a sufficiently large number).
3. Insert all vertices into a **Priority Queue** (or a similar data structure) based on their current distance.

### Pseudocode Representation
```markdown
FUNCTION Dijkstra(Graph, Source):
    Initialize DistanceArray with Infinity
    DistanceArray[Source] = 0
    Initialize PriorityQueue and insert all vertices with their distances
    
    WHILE PriorityQueue is not empty:
        // Extract the vertex with the smallest current distance
        CurrentVertex = PriorityQueue.ExtractMinimum()
        
        FOR EACH Neighbor of CurrentVertex:
            EdgeWeight = Weight(CurrentVertex, Neighbor)
            NewDistance = DistanceArray[CurrentVertex] + EdgeWeight
            
            // Check if the edge is tense
            IF NewDistance < DistanceArray[Neighbor]:
                // Relaxation Step
                DistanceArray[Neighbor] = NewDistance
                PredecessorArray[Neighbor] = CurrentVertex
                
                // Update the Neighbor's position in the Priority Queue
                PriorityQueue.DecreaseKey(Neighbor, NewDistance)
    
    RETURN DistanceArray
```

---

## 4. Implementation Details and Data Structures

### Priority Queue vs. Set
The algorithm frequently performs the `Extract Minimum` and `Decrease Key` operations.
* **C++ Priority Queue:** The standard `priority_queue` in C++ does not natively support the `Decrease Key` operation (updating an arbitrary value).
* **C++ Set:** A `std::set` is often preferred for implementation because it maintains a sorted collection and allows for the removal of arbitrary elements.

### Simulating "Decrease Key" with a Set
Since a Set doesn't have a direct "update" function, the operation is split into two parts:
1. **Delete** the old entry (vertex with old distance).
2. **Insert** the new entry (vertex with updated, smaller distance).

---

## 5. Summary of Main Takeaways

* **Core Purpose:** Dijkstra's algorithm solves the Single Source Shortest Path problem for graphs with **non-negative edge weights**.
* **Mechanism:** It operates similarly to BFS but uses a priority-based approach to account for weights, constantly "relaxing" edges to find the most efficient paths.
* **Constraints:** The algorithm is designed for positive weights; it may fail or produce incorrect results if the graph contains negative edge weights.
* **Efficiency:** By using efficient data structures like Heaps or Sets, the algorithm avoids the need for artificial vertex insertion used in simpler BFS modifications
