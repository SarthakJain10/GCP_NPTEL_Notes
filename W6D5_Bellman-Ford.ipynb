# Single Source Shortest Path (SSSP) with Negative Cycles: The Bellman-Ford Algorithm

This lecture focuses on solving the Single Source Shortest Path problem in graphs that contain negative edge weights and, more critically, negative weight cycles. It introduces the **Bellman-Ford Algorithm** as a robust solution for these scenarios.

---

## 1. Categorization of Edge Weights
The problem of finding the shortest path varies significantly depending on the nature of the edge weights:
* **Uniform/No Weights:** Solved efficiently using **Breadth-First Search (BFS)** in linear time.
* **Non-negative Weights:** Solved using **Dijkstra’s Algorithm**.
* **Negative Weights (No Negative Cycles):** Can be solved by a modified Dijkstra’s, though it may be computationally expensive.
* **Negative Cycles:** The standard shortest path is ill-defined; requires the **Bellman-Ford Algorithm**.

---

## 2. The Problem with Negative Cycles
A negative cycle is a path where the sum of the edge weights is less than zero.
* **Infinite Reduction:** If a path from a source to a destination passes through or can reach a negative cycle, a traveler could traverse that cycle infinitely many times.
* **Mathematical Result:** Each lap around the cycle reduces the total path cost. Consequently, there is no "shortest" path because you can always find a "shorter" one by looping again.
* **Algorithmic Goal:** In the presence of negative cycles, the objective shifts from finding a path to **detecting the cycle** and reporting it as a reason why a shortest path cannot be determined.

---

## 3. Core Concepts: Tense Edges and Relaxation
The Bellman-Ford algorithm relies on the concept of **relaxation**, similar to Dijkstra’s.

### The Tense Edge
An edge from vertex `u` to vertex `v` with weight `w` is considered **tense** if:
> `distance[u] + w < distance[v]`

This inequality indicates that a better path to `v` has been discovered by going through `u`.

### Relaxation
**Relaxing** an edge means updating the distance to `v` with the newly discovered shorter path:
> `distance[v] = distance[u] + w`

---

## 4. The Bellman-Ford Algorithm
Bellman-Ford is an iterative algorithm that repeatedly relaxes all edges in the graph.

### Algorithmic Logic
1.  **Initialization:** Set the distance to the source vertex to `0` and all other vertices to a very large number (representing infinity).
2.  **Main Loop:** Repeat the relaxation process for `N - 1` iterations (where `N` is the number of vertices).
3.  **Negative Cycle Detection:** Perform one final pass over all edges. If any edge is still "tense," a negative cycle exists.

### Pseudocode Representation
```markdown
function BellmanFord(Vertices, Edges, Source):
    // Step 1: Initialize distances
    for each vertex v in Vertices:
        distance[v] = INFINITY
    distance[Source] = 0

    // Step 2: Relax edges repeatedly
    // A simple path can have at most N-1 edges
    repeat N - 1 times:
        for each edge (u, v) with weight w in Edges:
            if distance[u] + w < distance[v]:
                distance[v] = distance[u] + w

    // Step 3: Check for negative weight cycles
    for each edge (u, v) with weight w in Edges:
        if distance[u] + w < distance[v]:
            return "Negative Cycle Detected"

    return distance
```


---

## 5. Performance and Complexity
The running time of the algorithm depends on how the graph is stored:
* **Adjacency List:** $O(N \times M)$, where $N$ is the number of vertices and $M$ is the number of edges.
* **Adjacency Matrix:** $O(N^3)$, as checking all possible edges in a matrix takes $O(N^2)$ per iteration.

---

## 6. Why N - 1 Iterations?
The lecture explains the logic behind the number of iterations:
* **Simple Paths:** In a graph without negative cycles, the shortest path between any two nodes is a **simple path** (no repeated vertices).
* **Edge Limit:** A simple path in a graph with $N$ vertices can contain at most $N - 1$ edges.
* **Convergence:** Each iteration of Bellman-Ford guarantees that it finds the shortest path for all paths consisting of *up to* $i$ edges. Therefore, by the $(N-1)^{th}$ iteration, all shortest simple paths are guaranteed to be found.
* **Cycle Proof:** If an improvement is still possible after $N - 1$ rounds, the path must contain more than $N - 1$ edges, implying a vertex is repeated and the path cost is still dropping—the definition of a negative cycle.

---

## Summary of Takeaways
* **Negative cycles** make the shortest path problem "ill-defined" because costs can be reduced infinitely.
* The **Bellman-Ford Algorithm** is the standard approach for Single Source Shortest Path when negative weights or cycles are possible.
* It functions by **relaxing every edge** in the graph exactly $N - 1$ times.
* Its primary utility is its ability to **detect negative cycles** by performing a $N^{th}$ check after the main loop finishes.
* While slower than Dijkstra's, it is more versatile as it handles negative edge weights that Dijkstra's cannot.
