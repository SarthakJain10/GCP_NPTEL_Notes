# Destroying Array: Problem Statement and Solution

This lecture covers a competitive programming problem titled **"Destroying Array"** (originally from a 2016 Codeforces elimination round). The core of the lecture focuses on understanding the problem mechanics and transitioning from a naive removal-based approach to an efficient solution using the **Disjoint Set Union (DSU)** data structure by reversing the process.


---

## 1. Problem Statement
The problem involves an array of $n$ positive integers and a specific sequence of "destruction" steps.

### Elements of the Problem:
* **Initial State:** An array containing $n$ numbers (e.g., $a_1, a_2, ..., a_n$).
* **Permutation of Destruction:** A sequence of indices representing the order in which elements are removed from the array.
* **The Process:** 1.  At each step, one element is removed based on the given permutation.
    2.  When an element is removed, the remaining elements form one or more **contiguous segments**.
* **The Goal:** After each destruction step, identify the **maximum sum** among all currently existing contiguous segments.

### Example Walkthrough:
If we have an array and we remove an element from the middle, the array splits into two separate contiguous segments. We must calculate the sum of elements in each segment and find the largest sum. This continues until the array is completely empty.

---

## 2. Algorithmic Approach: Reversing the Process
A direct simulation (removing elements) is difficult because "breaking" segments and recalculating sums is computationally expensive. 

### The Key Insight:
Instead of **destroying** the array, we can **reconstruct** it by processing the destruction sequence in **reverse order**.
* Start with an empty array.
* Add elements one by one in the reverse order of their destruction.
* As elements are added, they either form new segments or **merge** existing adjacent segments.
* This transformation turns a "breaking" problem into a "merging" problem, which is the primary use case for **Disjoint Set Union (DSU)**.

---

## 3. Implementation Logic with DSU
To track the segments and their sums efficiently, the DSU data structure is enhanced with "bookkeeping" for segment sums.

### Logic States for Adding an Element:
When an element at index `i` is added:
1.  **Isolated Element:** If neither neighbor (`i-1` or `i+1`) is present, the element forms a new singleton set.
2.  **Single Merge:** If only one neighbor is already present, the new element is merged into that neighbor's set.
3.  **Double Merge:** If both neighbors are already present, the new element acts as a bridge, merging the two existing segments into one large segment.

### Segment Sum Tracking:
* Each set in the DSU represents a contiguous segment.
* The **root** of each set stores the **total sum** of all elements in that segment.
* When two sets are merged, their sums are added together to update the new root.

---

## 4. Pseudocode
The following pseudocode outlines the reverse-reconstruction logic using DSU:

```python
FUNCTION SolveDestroyingArray(original_array, destruction_order):
    n = length(original_array)
    reversed_order = reverse(destruction_order)
    
    initialize DSU structure with n elements
    initialize segment_sums array with 0
    initialize exists_mask as FALSE for all indices
    
    current_max_sum = 0
    results_list = [0] // Final state sum is always 0
    
    FOR i FROM 0 TO n - 2: // Iterate through reversed order
        idx = reversed_order[i]
        exists_mask[idx] = TRUE
        segment_sums[idx] = original_array[idx]
        
        // Check left neighbor
        IF idx > 0 AND exists_mask[idx - 1] IS TRUE:
            Union(idx, idx - 1)
            
        // Check right neighbor
        IF idx < n - 1 AND exists_mask[idx + 1] IS TRUE:
            Union(idx, idx + 1)
            
        // Update global maximum sum
        root = Find(idx)
        new_segment_sum = GetSum(root)
        IF new_segment_sum > current_max_sum:
            current_max_sum = new_segment_sum
            
        results_list.APPEND(current_max_sum)
    
    RETURN reverse(results_list)
```

---

## 5. Summary of Main Takeaways
* **Reverse Thinking:** Many "deletion" or "destruction" problems in competitive programming can be simplified by reversing the operations to turn them into "addition" or "merging" problems.
* **DSU Versatility:** Beyond just tracking connectivity, DSU is excellent for maintaining aggregate data (like sums or sizes) across merging components.
* **Efficiency:** By using DSU with path compression and union by rank/size, the complexity remains nearly linear, which is far more efficient than re-scanning the array after every deletion.
