# Number Game Problem: Combinatorial Game Theory and the Golden Ratio

This lecture explores a "Numbers Game" originally featured in the Google Code Jam 2010. While it appears to be a complex game theory problem, it can be solved by identifying a specific mathematical threshold using observations and recursion.


## 1. Game Rules and Objectives
The game is played by two players, Arya and Bran, with two positive integers, **A** and **B**, written on a blackboard.

### Rules of Play
* **Turn-based**: Arya starts first, followed by Bran.
* **Legal Moves**: A player can update one of the numbers by subtracting a positive integer multiple ($k$) of the other number.
    * Update **A** to $A - (k \times B)$
    * OR Update **B** to $B - (k \times A)$
* **Constraint**: The resulting number must remain positive ($> 0$). Passing is not allowed.

### Winning and Losing Conditions
* **Losing**: The first player who is unable to make a move (i.e., any move would make a number zero or negative) loses the game .
* **Winning Position**: A state $(A, B)$ is a winning position if the current player has a strategy to win regardless of the opponent's moves .

## 2. Fundamental Observations
By analyzing simple cases, several patterns emerge regarding which positions are inherently winning or losing.

### Immediate Winning/Losing Scenarios
* **Identical Numbers $(A, A)$**: This is a **losing position** for the current player because any move results in a non-positive number .
* **Multiples $(A, k \times A)$**: If one number is a multiple of the other (where $k > 1$), it is a **winning position**. The player can subtract $(k-1)$ copies of $A$ to leave the opponent with $(A, A)$, which is a losing state .
* **One is 1 $(1, X)$**: If one number is 1 and the other is $X > 1$, the current player wins by reducing $X$ to 1, leaving $(1, 1)$ for the opponent .

### The "Forced Move" Range
If $B < A < 2B$, the current player has **only one valid move**: subtracting exactly one copy of $B$ from $A$ to get $(A-B, B)$. Any other $k$ would result in a non-positive value. This is a "forced" configuration .

## 3. The Recursive Strategy and the Golden Ratio
When $A \ge 2B$, the current player has multiple choices for $k$. This flexibility allows the player to choose a move that forces the opponent into a losing position .

### Expanding the Mystery Zone
Through recursive analysis, the "mystery zone" (where it isn't immediately obvious who wins) shrinks as follows:
* If $A \ge 2B$, the first player wins .
* Unraveling further layers of recursion shows thresholds related to ratios of Fibonacci numbers (e.g., $1.5, 1.6, 1.66\dots$) .

### The Golden Ratio ($\phi$) Threshold
The formal conclusion is that for a pair $(A, B)$ where $A > B$:
* The position is a **winning position** if and only if **$A \ge \phi \times B$**.
* The constant $\phi$ is the **Golden Ratio**, approximately **1.618** .

## 4. Problem Task: Counting Winning Pairs
The actual task is to count how many pairs $(A, B)$ are winning positions within given ranges:
* $A$ is in the range $[A1, A2]$
* $B$ is in the range $[B1, B2]$

### Optimization for Large Inputs
Since the ranges can be up to $1,000,000$, checking every pair $(A, B)$ individually is too slow ($O(N^2)$). Instead, we iterate through each $B$ and mathematically count the valid $A$ values.

## 5. Algorithmic Implementation (Pseudocode)

To solve the problem efficiently, we loop through the $B$ range and use the golden ratio threshold to count valid $A$ values in $O(N)$ time.

```markdown
// Initialize total winning count
Set TotalWinningPairs = 0
Set Phi = (1 + SquareRoot(5)) / 2

// Loop through each B in the range [B1, B2]
For each B from B1 to B2:
    
    // Case 1: A is the larger number (A >= Phi * B)
    Set LowerThresholdA = Phi * B
    // Count integers in [A1, A2] that are >= LowerThresholdA
    Set CountA = Max(0, A2 - Floor(LowerThresholdA))
    // Adjust if the entire range is above the threshold
    If A1 > LowerThresholdA:
        CountA = A2 - A1 + 1
    
    // Case 2: B is the larger number (B >= Phi * A)
    // This rearranges to A <= B / Phi (or A <= (Phi - 1) * B)
    Set UpperThresholdA = (Phi - 1) * B
    // Count integers in [A1, A2] that are <= UpperThresholdA
    Set CountB = Max(0, Floor(UpperThresholdA) - A1 + 1)
    // Adjust if the entire range is below the threshold
    If A2 < UpperThresholdA:
        CountB = A2 - A1 + 1

    // Add valid counts for this B to the running total
    TotalWinningPairs = TotalWinningPairs + CountA + CountB

Return TotalWinningPairs
```

## Summary of Takeaways
* **Combinatorial Games**: Many can be simplified by identifying "winning" and "losing" states through backward induction .
* **Threshold Discovery**: The "Numbers Game" reveals a deep connection to the **Golden Ratio ($\phi$)**; a position $(A, B)$ is winning if the ratio of the larger to the smaller number exceeds $\phi$.
* **Efficiency**: By fixing one variable ($B$) and using the mathematical threshold to count the other ($A$), we reduce the complexity from $O(N^2)$ to **$O(N)$**, making it feasible for inputs up to $10^6$
