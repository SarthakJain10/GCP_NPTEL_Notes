# Prim's Algorithm for Minimum Spanning Trees (MST)

This lecture covers the mechanics, simulation, and implementation of **Prim's Algorithm**, a greedy approach used to find the Minimum Spanning Tree of a weighted graph.

---

## 1. Core Concept and Framework

Prim’s algorithm is a greedy strategy that builds a spanning tree step-by-step by expanding a single connected component. It operates within the general framework of MST algorithms by making strategic use of **safe** and **useless** edges.


### Key Principles:
* **Starting Point**: The algorithm begins with an empty spanning forest where every vertex is isolated.
* **The Component (T)**: It picks an arbitrary starting vertex and designates it as the growing component **T**.
* **Greedy Selection**: In each step, the algorithm looks at all edges crossing the "cut" (edges with one endpoint inside **T** and the other outside). It selects the **cheapest** (minimum weight) edge.
* **Expansion**: The vertex at the other end of this cheapest edge is pulled into **T**.
* **Termination**: The process continues until all vertices are included, which occurs when the tree contains exactly **$n-1$** edges (where $n$ is the number of vertices).

---

## 2. Terminology and Logic

The lecture uses specific color-coding and terms to explain the logic of the algorithm:
* **Yellow (Visited Vertices)**: Vertices that have already been incorporated into the growing component **T**.
* **Red (Edges Under Consideration)**: Edges that have one endpoint in the "visited" set and one in the "unseen" set.
* **Green (Safe Edges)**: The specific edge chosen in a given step because it has the minimum weight among all red edges.
* **Useless Edges**: Edges where both endpoints are already inside **T**. These are ignored as they would create a cycle.

---

## 3. Algorithm Simulation Process

The simulation follows these logic steps:
1.  Initialize by marking a source vertex as "visited."
2.  Identify all edges connected to the source vertex and put them "under consideration."
3.  Among these, pick the one with the lowest weight.
4.  Add the new vertex at the end of that edge to the "visited" set.
5.  Update the edges under consideration by adding new edges incident to the newly added vertex.
6.  If an edge now has both endpoints in the "visited" set, it is discarded.
7.  Repeat until all vertices are visited or $n-1$ edges are selected.

*Note: If there is a tie between two edges of the same minimum weight, picking either will lead to a valid MST, though the final tree structure might differ.*

---

## 4. Implementation Details

The algorithm is implemented using a **Priority Queue** and an **Adjacency List** for efficiency. It shares structural similarities with Dijkstra’s algorithm but does not require edge relaxation.

### Data Structures:
* **Taken/Visited Array**: A boolean array to track which vertices are part of the MST.
* **Priority Queue (Min-Heap)**: Stores edges (weight and destination vertex) to always provide the cheapest edge currently available.
* **Adjacency List**: Stores the graph representation.

### Pseudocode Representation:

```PYTHON
FUNCTION PrimsAlgorithm(Graph, StartVertex):
    Initialize TakenArray with False
    Initialize MST_Cost to 0
    Initialize Edges_Count to 0
    Initialize MinHeap (Priority Queue)

    # Start with the first vertex
    Mark StartVertex as Taken
    FOR EACH neighbor, weight OF StartVertex:
        Push (weight, neighbor) INTO MinHeap

    WHILE MinHeap is NOT empty AND Edges_Count < (TotalVertices - 1):
        Extract (w, u) with smallest weight from MinHeap

        # If vertex u is already in MST, ignore this edge (Useless Edge)
        IF TakenArray[u] is True:
            CONTINUE

        # Otherwise, it's a safe edge
        Mark TakenArray[u] as True
        MST_Cost = MST_Cost + w
        Edges_Count = Edges_Count + 1

        # Add new edges incident to vertex u to the heap
        FOR EACH neighbor, weight OF u:
            IF TakenArray[neighbor] is False:
                Push (weight, neighbor) INTO MinHeap

    RETURN MST_Cost
END FUNCTION
```

---

## 5. Performance and Complexity

The efficiency of Prim's algorithm is dominated by the priority queue operations.

* **Priority Queue Operations**: Every edge is added (enqueued) and removed (dequeued) at most once.
* **Time Complexity**: $O(m \log n)$, where $m$ is the number of edges and $n$ is the number of vertices.
* **Comparison**: This is similar to Kruskal's algorithm in terms of performance. The choice between the two often depends on the specific problem or graph density.

---

## 6. Summary of Takeaways

* **Greedy Approach**: Prim's algorithm builds an MST by always choosing the cheapest local edge that connects a new vertex to the existing tree.
* **Correctness**: The algorithm is guaranteed to find the minimum possible cost for a spanning tree, even if multiple valid MSTs exist (in the case of weight ties).
* **Efficiency**: Using a **Min-Heap** ensures that we can quickly find the next safe edge, leading to a logarithmic complexity relative to the number of vertices.
* **Optimization**: An early exit is possible once $n-1$ edges are found, which is useful for large graphs or multiple test cases.
