# Implementation of the "Destroying Array" Problem Using DSU

This lecture focuses on the implementation details for the "Destroying Array" problem. The core strategy involves reversing the problem: instead of removing elements from an array (destruction), we add them in reverse order (reconstruction) using a **Disjoint Set Union (DSU)** data structure to track the sums of contiguous segments.

---

## 1. Data Structures and Initial Setup

The implementation requires several data structures to manage the state of the array and the sets of connected segments.

### Core Variables
* **Primary Array**: Stores the initial values of the elements.
* **State Array**: A boolean or integer array that tracks which locations are "occupied" (added back during reconstruction).
* **Sequence Vector**: Stores the order in which elements were destroyed (this will be reversed).
* **Answers Vector**: Stores the maximum segment sum found at each step of the reconstruction.

### Data Type Considerations
* **Integer (`int`)**: Used for the number of elements ($N$ up to $100,000$) and the indices of the array.
* **Long Long**: Essential for storing **segment sums** and the **final answers**. Since individual elements can be up to $10^9$, their sums will easily exceed the capacity of a standard 32-bit integer.

---

## 2. Reversing the Process

To use DSU effectively, the destruction process is handled in reverse:
1.  Start with an empty array.
2.  Process the destruction sequence from the last element to the first.
3.  Each step involves "adding" an element back and merging it with its existing neighbors to form or extend segments.

---

## 3. Disjoint Set Union (DSU) Modifications

A standard DSU implementation must be extended to handle the specific requirements of this problem.

### The `makeSet` Operation
Initially, the DSU is created with room for $N$ elements, but all sets are empty. When an element is added back, a `makeSet` operation is called for that specific index.

```pseudocode
Function makeSet(index, initialValue):
    Set parent[index] = index
    Set rank[index] = 0
    Set setSize[index] = 1
    Set segmentSum[index] = initialValue
    Increment totalNumberSets
```

### The `union` Operation
When merging two sets (segments), the sum of the elements in both sets must be combined.

```pseudocode
Function unionSets(index1, index2):
    root1 = find(index1)
    root2 = find(index2)
    
    If root1 is not equal to root2:
        // Use union by rank or size to attach root1 to root2
        Set parent[root1] = root2
        // CRITICAL: Update the sum of the new combined set
        Set segmentSum[root2] = segmentSum[root2] + segmentSum[root1]
```

---

## 4. The Reconstruction Algorithm

The heart of the solution lies in iterating through the reversed sequence and updating the "current maximum sum."

### Algorithm Logic
1.  **Reverse** the sequence of destruction.
2.  Initialize `currentMaximum` to 0.
3.  For each `elementIndex` in the reversed sequence:
    * Mark `elementIndex` as occupied in the **State Array**.
    * Perform `makeSet(elementIndex, valueAt[elementIndex])`.
    * **Check Left Neighbor**: If `elementIndex - 1` exists and is occupied, `unionSets(elementIndex, elementIndex - 1)`.
    * **Check Right Neighbor**: If `elementIndex + 1` exists and is occupied, `unionSets(elementIndex, elementIndex + 1)`.
    * **Update Champion**: Find the root of the newly formed segment and compare its `segmentSum` to `currentMaximum`. Update `currentMaximum` if the new sum is larger.
    * Store `currentMaximum` in the **Answers Vector**.

> **Note on Non-Negative Values**: Because all array elements are non-negative, adding an element or merging sets will never decrease the segment sum. Therefore, we only need to compare the newly updated segment's sum against the previous maximum.

---

## 5. Finalizing the Output

The **Answers Vector** collected during reconstruction contains the maximum sums in reverse chronological order of the original destruction.

* **Remove the last result**: The final step of reconstruction represents the full array sum, which is not needed (the problem asks for the state *after* a removal).
* **Reverse the Vector**: Flip the order of the answers to match the original destruction timeline.
* **Output**: Display the maximum segment sums.

---

## Summary of Takeaways

* **Offline Processing**: By reversing the destruction into a reconstruction, we can use DSU (which is designed for adding/merging) instead of trying to "split" sets.
* **Metadata Tracking**: DSU can track more than just connectivity; in this case, it maintains the **sum of elements** within each component.
* **Efficiency**: The approach leverages the nearly constant time complexity of DSU operations (using path compression and union by rank/size), making it efficient for $N = 100,000$.
* **Edge Cases**: Careful index management (1-based vs. 0-based) and boundary checks (neighbor existence) are vital for a correct implementation.
