# Disjoint Set Union (DSU) Implementation and Heuristics

This lecture focuses on the practical implementation of the **Disjoint Set Union (DSU)** data structure, specifically focusing on the `find` and `union` operations enhanced by heuristics like **Path Compression** and **Union by Rank**.


---

## 1. Class Structure and Initialization
The implementation follows an object-oriented approach to provide a clean and reusable interface. The data structure tracks parent pointers, ranks, and set sizes.

### Core Components
* **Parent Array (`p`):** Stores the parent pointer for each element. A leader element points to itself.
* **Rank Array (`rank`):** An upper bound on the depth of the tree, used to keep the tree balanced during unions.
* **Set Size Array (`setSize`):** Tracks the number of elements in each disjoint set.
* **Number of Sets (`numSets`):** A counter for the total number of distinct sets currently in the system.

### Initialization Logic
When initializing with $N$ elements, it is common to use $N+1$ indices to handle 1-based indexing scenarios easily.
* Each element $i$ is initially its own parent ($p[i] = i$).
* Ranks are initialized to $0$ as each tree consists of a single root.
* Set sizes are initialized to $1$.
* `numSets` is set to $N$.

---

## 2. The `Find` Operation with Path Compression
The `findSet` operation identifies the representative (leader) of the set containing element $i$.

### Path Compression Heuristic
To optimize future searches, the algorithm flattens the tree structure by making every node visited during the search point directly to the root.

### Pseudocode
```
Function findSet(i):
    If p[i] is equal to i:
        Return i
    Else:
        // Recursively find the root and update the parent pointer
        p[i] = findSet(p[i])
        Return p[i]
```

---

## 3. The `Union` Operation with Rank Heuristic
The `unionSet` operation merges two sets containing elements $i$ and $j$.

### Union by Rank Heuristic
To keep trees shallow, the algorithm always attaches the shorter tree (smaller rank) under the root of the taller tree (larger rank).

### Pseudocode
```
Function unionSet(i, j):
    If i and j are already in the same set:
        Return

    // Find representatives
    root_x = findSet(i)
    root_y = findSet(j)

    // Ensure root_x is the tree with the smaller rank
    If rank[root_x] > rank[root_y]:
        Swap(root_x, root_y)

    // Attach smaller tree to the root of the larger tree
    p[root_x] = root_y

    // Update rank if both trees had the same height
    If rank[root_x] == rank[root_y]:
        rank[root_y] = rank[root_y] + 1

    // Update bookkeeping
    setSize[root_y] = setSize[root_y] + setSize[root_x]
    numSets = numSets - 1
```

## 4. Helper and Auxiliary Functions
Additional functions enhance the utility of the DSU class for specific competitive programming problems.

* **`isSameSet(i, j)`:** Checks if two elements belong to the same set by comparing their representatives using `findSet`.
* **`sizeOfSet(i)`:** Returns the size of the set containing element $i$. Crucially, this must look up the size using the **leader** element's index.
* **`numberOfSets()`:** Returns the current count of disjoint sets.

### Handling Min/Max Metadata
For advanced queries (e.g., finding the smallest or largest element in a set), the class can maintain `minSet` and `maxSet` arrays.
* **Initialization:** `minSet[i] = i` and `maxSet[i] = i`.
* **Union Update:** When merging $root\_x$ into $root\_y$:
    * `minSet[root_y] = Minimum(minSet[root_y], minSet[root_x])`
    * `maxSet[root_y] = Maximum(maxSet[root_y], maxSet[root_x])`

---

## Summary of Key Takeaways
* **Efficiency:** Path compression and union by rank ensure that operations are nearly constant time.
* **Leader-Based Tracking:** Metadata like set size, rank, and min/max values are only guaranteed to be accurate for the **leader** element of a set.
* **Implementation Tip:** Initializing the structure with $N+1$ helps avoid indexing errors when problems use 1-to-$N$ numbering.
* **Versatility:** DSU is essential for problems involving connectivity, such as checking if two nodes are in the same component or tracking the growth of sets dynamically.
