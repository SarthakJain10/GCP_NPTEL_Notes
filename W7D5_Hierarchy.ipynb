# Graph Hierarchy and Minimum Spanning Trees

This lecture discusses a competitive programming problem from Codeforces titled **"Hierarchy,"** which involves building an organizational structure with minimum cost using principles from graph theory and **Minimum Spanning Trees (MST)**.

---

## 1. Problem Overview: The Company Hierarchy
The goal is to establish a supervisor-subordinate relationship among $N$ employees.

### Key Constraints and Rules
* **Tree Structure:** The hierarchy must form a tree.
* **Supervisor Limits:** Every employee except for one (the "CEO" or root) must have exactly **one** supervisor.
* **Applications:** There are $M$ applications where employee $A$ offers to supervise employee $B$ at a specific cost $C$.
* **Qualification Rule:** For an application to be valid, the qualification of the supervisor ($A$) must be strictly greater than the qualification of the subordinate ($B$).
* **Multiple Edges:** Different applications can exist for the same pair of employees with different costs.

### Objective
Calculate the **minimum cost** to build this hierarchy or determine if it is **impossible**.

---

## 2. Graph Modeling
To solve the problem, the employees and their relationships are modeled as a directed graph:
* **Vertices ($V$):** Represent the $N$ employees.
* **Directed Edges ($E$):** An edge from $A$ to $B$ with weight $C$ represents that $A$ can supervise $B$ for cost $C$.
* **In-degree Constraint:** In the final hierarchy, every vertex except the root must have an **in-degree of exactly one**.

### Identifying Impossible Scenarios
Building the hierarchy is impossible if:
1. **Multiple Potential Roots:** More than one employee has no one willing to supervise them.
2. **Disconnected Components:** The available applications do not allow for the formation of a single connected tree covering all $N$ employees.


---

## 3. Structural Properties
### Acyclic Nature
The hierarchy is naturally a **Directed Acyclic Graph (DAG)** because of the qualification constraint. Since a supervisor must have a strictly higher qualification than their subordinate, a cycle (e.g., $A$ supervises $B$, $B$ supervises $C$, and $C$ supervises $A$) is impossible as it would imply a person has a higher qualification than themselves.

### Connection to MST
While the problem involves directed edges, it is fundamentally a **Minimum Spanning Tree** problem because we need to select $N-1$ edges that connect all vertices with minimum total weight while satisfying the in-degree constraint.

---

## 4. Algorithmic Approaches
The lecture explores different ways to adapt standard MST algorithms for this problem.

### Choice of Algorithm: Kruskal’s Approach
The preferred method in this lecture is a modified version of **Kruskal’s Algorithm** using a **Union-Find (Disjoint Set Union)** data structure.

#### Modifications to Standard Kruskal’s:
1. **Sort Edges:** Sort all applications by cost in ascending order.
2. **In-degree Check:** When considering an edge $(U, V)$, ensure $V$ does not already have a supervisor. This is tracked by checking if $V$ is still its own parent in the hierarchy.
3. **Cycle Check:** Use the standard `find` operation to ensure $U$ and $V$ are not already in the same component to avoid cycles (though the qualification rule largely prevents this).

---

## 5. Pseudocode Implementation
The following pseudocode outlines the logic for solving the Hierarchy problem using the modified Kruskal's approach.

```python
# Modified Union-Find for Hierarchy
FUNCTION find(node):
    IF parent[node] == node:
        RETURN node
    ELSE:
        parent[node] = find(parent[node]) # Path compression
        RETURN parent[node]

FUNCTION union_hierarchy(supervisor, subordinate):
    root_supervisor = find(supervisor)
    root_subordinate = find(subordinate)
    
    IF root_supervisor != root_subordinate:
        # Assign supervisor directly to track hierarchy
        parent[subordinate] = supervisor
        RETURN True
    RETURN False

# Main Logic
PROCEDURE solve_hierarchy(N, applications):
    SORT applications by cost ASCENDING
    INITIALIZE total_cost = 0
    INITIALIZE edges_count = 0
    INITIALIZE parent array where parent[i] = i
    
    FOR EACH application (U, V, cost):
        # U is supervisor, V is subordinate
        # Standard MST cycle check + In-degree check
        IF find(U) != find(V) AND parent[V] == V:
            total_cost += cost
            edges_count += 1
            # Perform union specifically setting U as the supervisor
            parent[V] = U 
            
    IF edges_count == N - 1:
        PRINT total_cost
    ELSE:
        PRINT "Impossible"
```

## 6. Summary of Takeaways
* **Modeling:** Problems involving organizational structures can often be modeled as trees or directed graphs.
* **In-degree Constraint:** The requirement that "every person has exactly one supervisor" is equivalent to saying every node in a directed tree has an in-degree of one.
* **Adapting MST:** Standard algorithms like Kruskal's can be adapted by adding specific sanity checks (like the in-degree check) to fit the constraints of a directed graph.
* **Verification:** After running the greedy selection, always verify if the number of selected edges equals $N-1$ to ensure a complete hierarchy was formed.
